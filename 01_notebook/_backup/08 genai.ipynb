```
import os

from google.colab import userdata
import openai

from langchain_openai import ChatOpenAI
from langchain.schema import AIMessage, HumanMessage

import gradio as gr
```

# **1 | Übergreifende Parameter**
---



```
# OpenAI API Schlüssel setzen
OPENAI_API_KEY = userdata.get("OPENAI_API_KEY")
os.environ["OPENAI_API_KEY"] = OPENAI_API_KEY
```

# **2 | App-Integration ChatBot**
---



```
model = ChatOpenAI(model="gpt-4o-mini")


def predict(message, history):
    history_langchain_format = []
    for msg in history:
        if msg["role"] == "user":
            history_langchain_format.append(HumanMessage(content=msg["content"]))
        elif msg["role"] == "assistant":
            history_langchain_format.append(AIMessage(content=msg["content"]))
    history_langchain_format.append(HumanMessage(content=message))
    gpt_response = model.invoke(history_langchain_format)
    return gpt_response.content
```

# **3 | App Design**
---

```
demo = gr.ChatInterface(
    predict,
    type="messages",
    title="🤖 Einfacher ChatBot",
)
```

# **4 | Launch**
---

```
demo.launch()
```

# 2 | Funktionen
---

```
# Funktionen
def extract_text_with_pypdf2(pdf_path):
    text = ""
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        for page_num in range(len(reader.pages)):
            text += reader.pages[page_num].extract_text() + "\n"
    return text

def summarize_text(text, client):
    """Fasst Text mit der OpenAI API zusammen."""
    # Neues API-Format für Anfragen
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": "Fasse den folgenden Text präzise zusammen:"},
            {"role": "user", "content": text}
        ],
        max_tokens=1000
    )
    # Neue Antwortstruktur
    return response.choices[0].message.content.strip()
```

# 3 | LLM Summary

```
# Pfad zur PDF-Datei - englisch, ca. 40 Seiten
pdf_path = 'file.pdf'
```

```
# Initialisiere den OpenAI-Client
client = OpenAI()

# Text aus PDF extrahieren
extracted_text = extract_text_with_pypdf2(pdf_path)

# Text zusammenfassen
summary = summarize_text(extracted_text, client)

# Text ausgeben
display(Markdown("## Zusammenfassung"))
display(Markdown("---"))
display(Markdown(summary))
```
