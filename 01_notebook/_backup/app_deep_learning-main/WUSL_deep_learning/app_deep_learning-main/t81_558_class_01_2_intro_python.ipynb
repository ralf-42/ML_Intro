Variablen ermöglichen Ihrem Programm, Werte zu speichern, die sich während der Ausführung des Programms ändern können. Variablen haben Namen, die es Ihnen ermöglichen, auf ihre Werte zu verweisen. Der folgende Code weist einer Variablen namens „a“ einen Integerwert und einer Variablen namens „b“ einen Zeichenfolgenwert zu.

Sie können Zeichenfolgen und Variablen zum Drucken mischen. Diese Technik wird als formatierte oder interpolierte Zeichenfolge bezeichnet. Die Variablen müssen sich innerhalb der geschweiften Klammern befinden. In Python wird diese Art von Zeichenfolge im Allgemeinen als F-Zeichenfolge bezeichnet. Die F-Zeichenfolge wird gekennzeichnet, indem ein „f“ direkt vor das öffnende einfache oder doppelte Anführungszeichen gesetzt wird, mit dem die Zeichenfolge beginnt. Der folgende Code demonstriert die Verwendung einer F-Zeichenfolge, um mehrere Variablen mit einer Literalzeichenfolge zu mischen.


```python
a = 10
print(f"Der Wert von a ist {a}")
```

Sie können F-Strings auch mit mathematischen Ausdrücken verwenden (Ausdrücke genannt). Geschweifte Klammern können jeden gültigen Python-Ausdruck zum Drucken einschließen. Der folgende Code demonstriert die Verwendung eines Ausdrucks innerhalb der geschweiften Klammern eines F-Strings.


```python
a = 10
print(f"Der Wert von a plus 5 ist {a+5}")
```

Python bietet viele Möglichkeiten, Zahlen auszugeben. Diese sind alle korrekt. Für diesen Kurs verwenden wir jedoch F-Strings. Der folgende Code demonstriert einige der verschiedenen Methoden zum Drucken von Zahlen in Python.


```python
a = 5

print(f"a is {a}")  # Bevorzugte Methode für diesen Kurs.
print("a is {}".format(a))
print("a is " + str(a))
print("a is %d" % (a))
```

Sie können if-Anweisungen verwenden, um Logik auszuführen. Sehen Sie die Einrückungen? Mit diesen if-Anweisungen definiert Python Codeblöcke, die zusammen ausgeführt werden sollen. Ein Block beginnt normalerweise nach einem Doppelpunkt und umfasst alle Zeilen auf derselben Einrückungsebene. Im Gegensatz zu vielen anderen Programmiersprachen verwendet Python Leerzeichen, um Codeblöcke zu definieren. Die Tatsache, dass Leerzeichen für die Bedeutung des Programmcodes von Bedeutung sind, ist für neue Python-Programmierer häufig eine Quelle der Verärgerung. Tabulatoren und Leerzeichen werden beide verwendet, um den Gültigkeitsbereich in einem Python-Programm zu definieren. Es wird nicht empfohlen, sowohl Leerzeichen als auch Tabulatoren im selben Programm zu verwenden.


```python
a = 5
if a > 5:
    print("The variable a is greater than 5.")
else:
    print("The variable a is not greater than 5")
```

Die folgende if-Anweisung hat mehrere Ebenen. Es kann leicht passieren, dass diese Ebenen falsch eingerückt werden, seien Sie also vorsichtig. Dieser Code enthält eine verschachtelte if-Anweisung unter der ersten if-Anweisung „a==5“. Nur wenn a gleich 5 ist, wird die verschachtelte if-Anweisung „b==6“ ausgeführt. Beachten Sie auch, dass der Befehl „elif“ „else if“ bedeutet.


```python
a = 5
b = 6

if a == 5:
    print("The variable a is 5")
    if b == 6:
        print("The variable b is also 6")
elif a == 6:
    print("The variable a is 6")
```

Es ist auch wichtig zu beachten, dass der doppelte Gleichheitsoperator ("==") verwendet wird, um die Gleichheit zweier Ausdrücke zu testen. Der einfache Gleichheitsoperator ("=") wird nur verwendet, um Variablen in Python Werte zuzuweisen. Die Operatoren "größer als" ("">"), "kleiner als" ("<"), "größer als oder gleich" ("">=") und "kleiner als oder gleich" ("<=") funktionieren alle so, wie es allgemein akzeptiert wird. Die Prüfung auf Ungleichheit wird mit dem Operator "ungleich" ("!=") durchgeführt.

In Programmiersprachen ist es üblich, eine Schleife über einen Zahlenbereich auszuführen. Python erreicht dies durch die Verwendung der **range**-Operation. Hier sehen Sie eine **for**-Schleife und eine **range**-Operation, die das Programm zwischen 1 und 3 laufen lässt.


```python
for x in range(1, 3):  # Wenn Sie xrange jemals sehen, sind Sie in Python 2
    print(x)
# Wenn Sie jemals print x (ohne Klammern) sehen, sind Sie in Python 2
```

Dieser Code veranschaulicht einige Inkompatibilitäten zwischen Python 2 und Python 3. Vor Python 3 war es zulässig, die Klammern bei einem Funktionsaufruf _print_ wegzulassen. Diese Methode zum Aufrufen des Befehls _print_ ist in Python 3 nicht mehr zulässig. Ebenso war es manchmal eine Leistungsverbesserung, den Befehl _xrange_ anstelle des Befehls range zu verwenden. Python 3 hat die gesamte Funktionalität des Python 2-Befehls _xrange_ in den normalen Befehl _range_ integriert. Daher sollte der Programmierer den Befehl _xrange_ in Python 3 nicht verwenden. Wenn Sie eine dieser Konstrukte im Beispielcode sehen, sehen Sie sich ein älteres Beispiel aus der Python 2-Ära an.

Der Befehl _range_ wird in Verbindung mit Schleifen verwendet, um einen bestimmten Zahlenbereich zu durchlaufen. Fälle, in denen Sie über bestimmte Zahlenbereiche schleifen müssen, sind eher ungewöhnlich. Im Allgemeinen verwenden Programmierer Schleifen für Sammlungen von Elementen, anstatt numerische Werte fest in Ihren Code zu Kodieren. Sammlungen sowie die Operationen, die Schleifen an ihnen ausführen können, werden später in diesem Modul behandelt.

Nachfolgend sehen Sie ein weiteres Beispiel für die Schleifenausgabe von Zeichenfolgen und Zahlen.


```python
acc = 0
for x in range(1, 3):
    acc += x
print(f"Beim Hinzufügen von {x} beträgt die Summe bis jetzt {acc}")

print(f"Endgültige Summe: {acc}")
```
