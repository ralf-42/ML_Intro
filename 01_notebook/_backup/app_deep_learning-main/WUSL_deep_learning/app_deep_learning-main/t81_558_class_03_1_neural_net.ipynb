Neuronale Netzwerke waren eines der ersten Modelle des maschinellen Lernens. Ihre Popularität ist zweimal gesunken und steigt nun zum dritten Mal. Deep Learning impliziert die Verwendung neuronaler Netzwerke. Das „Deep“ in Deep Learning bezieht sich auf ein neuronales Netzwerk mit vielen verborgenen Schichten. Da es neuronale Netzwerke schon so lange gibt, sind sie ziemlich altmodisch. Forscher haben viele verschiedene Trainingsalgorithmen, Aktivierungs-/Übertragungsfunktionen und Strukturen entwickelt. Neuronale Netzwerke akzeptieren Eingaben und erzeugen Ausgaben. Die Eingabe in ein neuronales Netzwerk wird als Merkmalsvektor bezeichnet. Die Größe dieses Vektors hat immer eine feste Länge. Wenn Sie die Größe des Merkmalsvektors ändern, bedeutet dies normalerweise, dass das gesamte neuronale Netzwerk neu erstellt werden muss. Obwohl der Merkmalsvektor als „Vektor“ bezeichnet wird, ist dies nicht immer der Fall. Ein Vektor impliziert ein 1D-Array. Später werden wir etwas über Convolutional Neural Networks (CNNs) lernen, bei denen die Eingabegröße geändert werden kann, ohne dass das neuronale Netzwerk neu trainiert werden muss. Historisch gesehen war die Eingabe in ein neuronales Netzwerk immer 1D. Bei modernen neuronalen Netzwerken sehen Sie jedoch möglicherweise Eingabedaten wie:

- **1D-Vektor** – Klassische Eingabe für ein neuronales Netzwerk, ähnlich den Zeilen in einer Kalkulationstabelle. Häufig bei prädiktiver Modellierung.
- **2D-Matrix** – Graustufenbild-Eingabe für ein CNN.
- **3D-Matrix** – Farbbildeingabe für ein CNN.
- **nD-Matrix** – Höherwertige Eingabe für ein CNN.

Vor CNNs kodierten Programme Bilder entweder in eine Zwischenform oder schickten die Bildeingabe an ein neuronales Netzwerk, indem sie die Bildmatrix einfach in ein langes Array komprimierten, indem sie die Bildzeilen nebeneinander platzierten. CNNs unterscheiden sich, da die Matrix die Schichten des neuronalen Netzwerks durchläuft.

Anfangs konzentriert sich dieses Buch auf 1D-Eingaben für neuronale Netzwerke. Spätere Module konzentrieren sich jedoch stärker auf Eingaben höherer Dimensionen.

Der Begriff Dimension kann bei neuronalen Netzwerken verwirrend sein. Im Sinne eines 1D-Eingabevektors bezieht sich die Dimension darauf, wie viele Elemente sich in diesem 1D-Array befinden. Beispielsweise hat ein neuronales Netzwerk mit zehn Eingabeneuronen zehn Dimensionen. Da wir jetzt jedoch CNNs haben, hat die Eingabe Dimensionen. Die Eingabe für das neuronale Netzwerk hat _normalerweise_ 1, 2 oder 3 Dimensionen. Vier oder mehr Dimensionen sind ungewöhnlich. Sie könnten eine 2D-Eingabe für ein neuronales Netzwerk mit 64 x 64 Pixeln haben. Diese Konfiguration würde zu 4.096 Eingabeneuronen führen. Dieses Netzwerk ist entweder 2D oder 4.096D, je nachdem, auf welche Dimensionen Sie sich beziehen.

## Neuronen und Schichten

Die meisten neuronalen Netzwerkstrukturen verwenden irgendeine Art von Neuron. Es gibt viele verschiedene neuronale Netzwerke, und Programmierer führen experimentelle neuronale Netzwerkstrukturen ein. Daher ist es nicht möglich, jede neuronale Netzwerkarchitektur abzudecken. Es gibt jedoch einige Gemeinsamkeiten zwischen neuronalen Netzwerkimplementierungen. Ein neuronaler Netzwerkalgorithmus besteht normalerweise aus einzelnen, miteinander verbundenen Einheiten, auch wenn diese Einheiten Neuronen genannt werden können oder nicht. Der Name für eine neuronale Netzwerkverarbeitungseinheit variiert in den Literaturquellen. Sie könnte Knoten, Neuron oder Einheit genannt werden.

Ein Diagramm zeigt die abstrakte Struktur eines einzelnen künstlichen Neurons in Abbildung 3.ANN.

**Abbildung 3.ANN: Ein künstliches Neuron**
![An Artificial Neuron](https://data.heatonresearch.com/images/wustl/class/class_2_abstract_nn.png "An Artificial Neuron")

Das künstliche Neuron empfängt Eingaben von einer oder mehreren Quellen, bei denen es sich um andere Neuronen oder um Daten handeln kann, die von einem Computerprogramm in das Netzwerk eingespeist werden. Diese Eingaben sind normalerweise Gleitkommazahlen oder Binärzahlen. Binäre Eingaben werden häufig in Gleitkommazahlen kodiert, indem Wahr oder Falsch als 1 oder 0 dargestellt wird. Manchmal stellt das Programm die binären Informationen auch in einem bipolaren System dar, wobei Wahr als Eins und Falsch als -1 dargestellt wird.

Ein künstliches Neuron multipliziert jeden dieser Inputs mit einem Gewicht. Dann addiert es diese Multiplikationen und übergibt diese Summe an eine Aktivierungsfunktion. Einige neuronale Netzwerke verwenden keine Aktivierungsfunktion. Die folgende Gleichung fasst den berechneten Output eines Neurons zusammen:

$$ f(x,w) = \phi(\sum_i(\theta_i \cdot x_i)) $$

In der obigen Gleichung stellen die Variablen $x$ und $\theta$ die Eingabe und Gewichte des Neurons dar. Die Variable $i$ entspricht der Anzahl der Gewichte und Eingaben. Sie müssen immer die gleiche Anzahl an Gewichten wie Eingaben haben. Das neuronale Netzwerk multipliziert jedes Gewicht mit seiner jeweiligen Eingabe und speist die Produkte dieser Multiplikationen in eine Aktivierungsfunktion ein, die mit dem griechischen Buchstaben $\phi$ (phi) gekennzeichnet ist. Dieser Prozess führt zu einer einzigen Ausgabe des Neurons.

## Arten von Neuronen

Im letzten Abschnitt haben wir kurz die Idee vorgestellt, dass es verschiedene Arten von Neuronen gibt. Nicht jedes neuronale Netzwerk verwendet jede Art von Neuron. Es ist auch möglich, dass ein einzelnes Neuron die Rolle mehrerer verschiedener Neuronentypen übernimmt. Jetzt werden wir alle im Kurs beschriebenen Neuronentypen erklären.

In einem neuronalen Netzwerk gibt es normalerweise vier Arten von Neuronen:

- **Eingabeneuronen** – Wir ordnen jedes Eingabeneuron einem Element im Merkmalsvektor zu.
- **Versteckte Neuronen** – Versteckte Neuronen ermöglichen es dem neuronalen Netzwerk, abstrakt zu sein und die Eingabe in die Ausgabe zu verarbeiten.
- **Ausgabeneuronen** – Jedes Ausgabeneuron berechnet einen Teil der Ausgabe.
- **Bias-Neuronen** – Funktionieren ähnlich wie der y-Achsenabschnitt einer linearen Gleichung.

Wir platzieren jedes Neuron in einer Schicht:

- **Eingabeebene** - Die Eingabeebene akzeptiert Merkmalsvektoren aus dem Datensatz. Eingabeebenen haben normalerweise ein Biasneuron.
- **Ausgabeebene** - Die Ausgabe des neuronalen Netzwerks. Die Ausgabeebene hat kein Biasneuron.
- **Versteckte Schichten** - Schichten zwischen den Eingabe- und Ausgabeschichten. Jede versteckte Schicht hat normalerweise ein Bias-Neuron.

## Eingangs- und Ausgangsneuronen

Fast jedes neuronale Netzwerk hat Eingangs- und Ausgangsneuronen. Die Eingangsneuronen akzeptieren Daten vom Programm für das Netzwerk. Das Ausgangsneuron liefert verarbeitete Daten vom Netzwerk zurück an das Programm. Das Programm gruppiert diese Eingangs- und Ausgangsneuronen in separate Schichten, die als Eingangs- und Ausgangsschichten bezeichnet werden. Das Programm stellt die Eingabe für ein neuronales Netzwerk normalerweise als Array oder Vektor dar. Die Anzahl der im Vektor enthaltenen Elemente muss der Anzahl der Eingangsneuronen entsprechen. Beispielsweise könnte ein neuronales Netzwerk mit drei Eingangsneuronen den folgenden Eingangsvektor akzeptieren:

$$ [0,5, 0,75, 0,2] $$

Neuronale Netzwerke akzeptieren normalerweise Gleitkommavektoren als Eingabe. Der Konsistenz halber stellen wir die Ausgabe eines einzelnen Ausgabeneuronnetzwerks als einelementigen Vektor dar. Ebenso geben neuronale Netzwerke einen Vektor mit einer Länge aus, die der Anzahl der Ausgabeneuronen entspricht. Die Ausgabe ist häufig ein einzelner Wert von einem einzelnen Ausgabeneuron.

## Versteckte Neuronen

Versteckte Neuronen haben zwei wesentliche Eigenschaften. Erstens erhalten versteckte Neuronen nur Eingaben von anderen Neuronen, wie z. B. Eingabe- oder andere versteckte Neuronen. Zweitens geben versteckte Neuronen nur an andere Neuronen aus, wie z. B. Ausgabe- oder andere versteckte Neuronen. Versteckte Neuronen helfen dem neuronalen Netzwerk, die Eingabe zu verstehen und die Ausgabe zu bilden. Programmierer gruppieren versteckte Neuronen oft in vollständig verbundene versteckte Schichten. Diese versteckten Schichten verarbeiten jedoch die eingehenden Daten oder die endgültige Ausgabe nicht direkt.

Eine häufige Frage von Programmierern betrifft die Anzahl der verborgenen Neuronen in einem Netzwerk. Da die Antwort auf diese Frage komplex ist, wird die Anzahl der verborgenen Neuronen in mehreren Abschnitten des Kurses behandelt. Vor dem Deep Learning meinten Forscher im Allgemeinen, dass alles, was mehr als eine einzige verborgene Schicht ist, übertrieben sei. [[Cite:hornik1989multilayer]](https://www.sciencedirect.com/science/article/abs/pii/0893608089900208) Forscher haben bewiesen, dass ein neuronales Netzwerk mit einer einzigen verborgenen Schicht als universeller Approximator fungieren kann. Mit anderen Worten: Dieses Netzwerk sollte lernen können, aus jeder Eingabe jede Ausgabe zu erzeugen (oder anzunähern), solange es über genügend verborgene Neuronen in einer einzigen Schicht verfügt.

Training bezeichnet den Prozess, der gute Gewichtswerte bestimmt. Vor dem Aufkommen von Deep Learning befürchteten Forscher, dass zusätzliche Schichten die Trainingszeit verlängern oder zu Überanpassung führen würden. Beide Bedenken sind berechtigt; höhere Hardwaregeschwindigkeiten und clevere Techniken können diese Bedenken jedoch ausräumen. Bevor Forscher Deep Learning-Techniken einführten, hatten wir keine effiziente Möglichkeit, ein tiefes Netzwerk zu trainieren, also ein neuronales Netzwerk mit vielen verborgenen Schichten. Obwohl ein neuronales Netzwerk mit einer einzigen verborgenen Schicht theoretisch alles lernen kann, ermöglicht Deep Learning eine komplexere Darstellung von Mustern in den Daten.

## Bias-Neuronen

Programmierer fügen neuronalen Netzwerken Bias-Neuronen hinzu, um ihnen beim Erlernen von Mustern zu helfen. Bias-Neuronen funktionieren wie ein Eingangsneuron, das immer den Wert 1 erzeugt. Da die Bias-Neuronen einen konstanten Ausgang von 1 haben, sind sie nicht mit der vorherigen Schicht verbunden. Der Wert 1, die sogenannte Bias-Aktivierung, kann auf andere Werte als 1 eingestellt werden. 1 ist jedoch die häufigste Bias-Aktivierung. Nicht alle neuronalen Netzwerke haben Bias-Neuronen. ## Andere Neuronentypen

Die einzelnen Einheiten, aus denen ein neuronales Netzwerk besteht, werden nicht immer als Neuronen bezeichnet. Forscher bezeichnen diese Neuronen manchmal als Knoten, Einheiten oder Summationen. ## Warum werden Bias-Neuronen benötigt?

Die Aktivierungsfunktionen aus dem vorherigen Abschnitt geben die Ausgabe eines einzelnen Neurons an. Gemeinsam formen das Gewicht und die Tendenz eines Neurons die Ausgabe der Aktivierung, um die gewünschte Ausgabe zu erzeugen. Um zu sehen, wie dieser Prozess abläuft, betrachten Sie die folgende Gleichung. Sie stellt ein neuronales Netzwerk mit Sigmoid-Aktivierung und einem einzigen Eingang dar.

$$ f(x,w,b) = \frac{1}{1 + e^{-(wx+b)}} $$

Die Variable $x$ stellt den einzelnen Input für das neuronale Netzwerk dar. Die Variablen $w$ und $b$ geben das Gewicht und die Vorspannung des neuronalen Netzwerks an. Die obige Gleichung kombiniert die gewichtete Summe der Inputs und die Sigmoid-Aktivierungsfunktion. In diesem Abschnitt werden wir die Sigmoid-Funktion betrachten, da sie die Wirkung eines Bias-Neurons demonstriert.

Mit den Gewichten des Neurons können Sie die Steigung oder Form der Aktivierungsfunktion anpassen. Abbildung 3.A-WEIGHT zeigt die Auswirkung auf die Ausgabe der Sigmoid-Aktivierungsfunktion, wenn das Gewicht variiert wird:

**Abbildung 3.A-GEWICHT: Neuronengewichtsverlagerung**
![Adjusting Weight](https://data.heatonresearch.com/images/wustl/class/class_2_bias_weight.png "Neuron Weight Shifting")

Das obige Diagramm zeigt mehrere Sigmoidkurven mit den folgenden Parametern:

$$ f(x,0,5,0,0) $$
$$ f(x,1,0,0,0) $$
$$ f(x,1,5,0,0) $$
$$ f(x,2.0,0.0) $$

Wir haben keine Verzerrung verwendet, um die Kurven zu erzeugen, was sich in dem dritten Parameter von 0 in jedem Fall zeigt. Die Verwendung von vier Gewichtungswerten ergibt vier verschiedene Sigmoidkurven in der obigen Abbildung. Unabhängig von der Gewichtung erhalten wir immer denselben Wert von 0,5, wenn _x_ 0 ist, da alle Kurven denselben Punkt erreichen, wenn x 0 ist. Möglicherweise benötigen wir das neuronale Netzwerk, um andere Werte zu erzeugen, wenn der Input nahe 0,5 liegt.

Der Bias verschiebt die Sigmoidkurve, was andere Werte als 0,5 zulässt, wenn _x_ nahe 0 liegt. Abbildung 3.A-BIAS zeigt die Auswirkung der Verwendung eines Gewichts von 1,0 mit mehreren unterschiedlichen Biasen:

**Abbildung 3.A-BIAS: Neuronale Bias-Verschiebung**
![Adjusting Bias](https://data.heatonresearch.com/images/wustl/class/class_2_bias_value.png "Neuron Bias Shifting")

Das obige Diagramm zeigt mehrere Sigmoidkurven mit den folgenden Parametern:

$$ f(x,1.0,1.0) $$
$$ f(x,1,0,0,5) $$
$$ f(x,1,0,1,5) $$
$$ f(x,1,0,2,0) $$

## Moderne Aktivierungsfunktionen

Aktivierungsfunktionen, auch als Übertragungsfunktionen bekannt, werden verwendet, um die Ausgabe jeder Schicht eines neuronalen Netzwerks zu berechnen. In der Vergangenheit verwendeten neuronale Netzwerke eine hyperbolische Tangens-, Sigmoid-/Logistik- oder lineare Aktivierungsfunktion. Moderne tiefe neuronale Netzwerke verwenden jedoch hauptsächlich die folgenden Aktivierungsfunktionen:

- **Rectified Linear Unit (ReLU)** – Wird für die Ausgabe von versteckten Ebenen verwendet. [[Cite:glorot2011deep]](http://proceedings.mlr.press/v15/glorot11a/glorot11a.pdf)
- **Softmax** – Wird für die Ausgabe von Klassifizierungs-Neuronalen Netzwerken verwendet.
- **Linear** – Wird für die Ausgabe von Regressionsneuronalen Netzwerken (oder 2-Klassen-Klassifizierung) verwendet.

## Lineare Aktivierungsfunktion

Die grundlegendste Aktivierungsfunktion ist die lineare Funktion, da sie die Neuronenausgabe nicht ändert. Die folgende Gleichung 1.2 zeigt, wie das Programm normalerweise eine lineare Aktivierungsfunktion implementiert:

$$ \phi(x) = x $$

Wie Sie sehen, gibt diese Aktivierungsfunktion einfach den Wert zurück, den die Neuroneneingaben an sie übergeben haben. Abbildung 3.LIN zeigt den Graphen für eine lineare Aktivierungsfunktion:

**Abbildung 3.LIN: Lineare Aktivierungsfunktion**
![Linear Activation Function](https://data.heatonresearch.com/images/wustl/class/graphs-linear.png "Linear Activation Function")

Regressionsneuronale Netze, die lernen, numerische Werte bereitzustellen, verwenden in der Regel eine lineare Aktivierungsfunktion in ihrer Ausgabeschicht. Klassifizierungsneuronale Netze, die eine geeignete Klasse für ihre Eingabe bestimmen, verwenden in ihrer Ausgabeschicht häufig eine Softmax-Aktivierungsfunktion.

## Gleichgerichtete lineare Einheiten (ReLU)

Die meisten aktuellen Forschungsergebnisse empfehlen jetzt ReLU aufgrund besserer Trainingsergebnisse. Daher sollten die meisten neuronalen Netzwerke ReLU auf verborgenen Ebenen und entweder Softmax oder linear auf der Ausgabeebene verwenden. Die folgende Gleichung zeigt die einfache ReLU-Funktion:

## Softmax-Aktivierungsfunktion

Die letzte Aktivierungsfunktion, die wir untersuchen werden, ist die Softmax-Aktivierungsfunktion. Neben der linearen Aktivierungsfunktion finden Sie die Softmax-Funktion normalerweise in der Ausgabeschicht eines neuronalen Netzwerks. Klassifizierungsneuronale Netzwerke verwenden normalerweise die Softmax-Funktion. Das Neuron mit dem höchsten Wert beansprucht die Eingabe als Mitglied seiner Klasse. Da dies eine bevorzugte Methode ist, zwingt die Softmax-Aktivierungsfunktion die Ausgabe des neuronalen Netzwerks, die Wahrscheinlichkeit darzustellen, dass die Eingabe in jede der Klassen fällt. Die Ausgaben des Neurons sind numerische Werte ohne Softmax, wobei der höchste Wert die Gewinnerklasse angibt.

Um zu sehen, wie das Programm die Softmax-Aktivierungsfunktion verwendet, sehen wir uns ein typisches Klassifizierungsproblem eines neuronalen Netzwerks an. Der Iris-Datensatz enthält vier Messungen für 150 verschiedene Irisblüten. Jede dieser Blüten gehört zu einer von drei Irisarten. Wenn Sie die Messungen einer Blüte angeben, kann das neuronale Netzwerk Ihnen mithilfe der Softmax-Funktion die Wahrscheinlichkeit nennen, mit der diese Messungen zu jeder der drei Arten gehören. Das neuronale Netzwerk könnte Ihnen beispielsweise sagen, dass die Wahrscheinlichkeit, dass es sich bei der Iris um eine Setosa-Blüte handelt, 80 % beträgt, dass es sich um eine Virginica-Blüte handelt, 15 % und dass es sich nur um eine Versicolor-Blüte handelt, nur 5 %. Da es sich dabei um Wahrscheinlichkeiten handelt, müssen sie zusammen 100 % ergeben. Es könnte nicht eine Wahrscheinlichkeit von 80 % für Setosa, 75 % für Virginica und 20 % für Versicolor geben – ein solches Ergebnis wäre unsinnig.

Um Eingabedaten einer von drei Irisarten zuzuordnen, benötigen Sie für jede Art ein Ausgabeneuron. Die Ausgabeneuronen geben nicht automatisch die Wahrscheinlichkeit jeder der drei Arten an. Daher ist es wünschenswert, Wahrscheinlichkeiten bereitzustellen, die sich zu 100 % summieren. Das neuronale Netzwerk gibt die Wahrscheinlichkeit an, mit der eine Blume zu jeder der drei Arten gehört. Um die Wahrscheinlichkeit zu ermitteln, verwenden Sie die Softmax-Funktion in der folgenden Gleichung:

$$ \phi*i(x) = \frac{exp(x_i)}{\sum*{j}^{ }exp(x_j)} $$

In der obigen Gleichung stellt $i$ den Index des Ausgabeneurons ($\phi$) dar, den das Programm berechnet, und $j$ stellt die Indizes aller Neuronen in der Gruppe/Ebene dar. Die Variable $x$ bezeichnet das Array der Ausgabeneuronen. Es ist wichtig zu beachten, dass das Programm die Softmax-Aktivierung anders berechnet als die anderen Aktivierungsfunktionen in diesem Modul. Wenn Softmax die Aktivierungsfunktion ist, ist die Ausgabe eines einzelnen Neurons von den anderen Ausgabeneuronen abhängig.

Um die Softmax-Funktion in Aktion zu sehen, beziehen Sie sich auf diesen [Softmax example website](http://www.heatonresearch.com/aifh/vol3/softmax.html).

Stellen Sie sich ein trainiertes neuronales Netzwerk vor, das Daten in drei Kategorien einteilt: die drei Irisarten. In diesem Fall würden Sie für jede der Zielklassen ein Ausgabeneuron verwenden. Stellen Sie sich vor, das neuronale Netzwerk würde Folgendes ausgeben:

- **Neuron 1**: setosa: 0,9
- **Neuron 2**: versicolor: 0,2
- **Neuron 3**: virginica: 0,4

Die obige Ausgabe zeigt, dass das neuronale Netzwerk die Daten als eine Setosa-Schwertlilie betrachtet. Diese Zahlen sind jedoch keine Wahrscheinlichkeiten. Der Wert 0,9 stellt keine 90-prozentige Wahrscheinlichkeit dar, dass die Daten eine Setosa darstellen. Die Summe dieser Werte beträgt 1,5. Damit das Programm sie als Wahrscheinlichkeiten behandelt, müssen sie 1,0 ergeben. Der Ausgabevektor für dieses neuronale Netzwerk ist der folgende:

$$ [0,9,0,2,0,4] $$

Wenn Sie der Softmax-Funktion diesen Vektor übergeben, wird der folgende Vektor zurückgegeben:

$$ [0,47548495534876745, 0,2361188410001125, 0,28839620365112] $$

Die Summe der drei oben genannten Werte beträgt 1,0 und kann als Wahrscheinlichkeit betrachtet werden. Die Wahrscheinlichkeit, dass die Daten eine Setosa-Schwertlilie darstellen, beträgt 48 %, da der erste Wert im Vektor auf 0,48 (48 %) gerundet wird. Sie können diesen Wert folgendermaßen berechnen:

$$ Summe=\exp(0,9)+\exp(0,2)+\exp(0,4)=5,17283056695839 $$
$$ j_0= \exp(0,9)/Summe = 0,47548495534876745 $$
$$ j_1= \exp(0,2)/Summe = 0,2361188410001125 $$
$$ j_2= \exp(0,4)/Summe = 0,28839620365112 $$

## Schrittaktivierungsfunktion

Die Schritt- oder Schwellenaktivierungsfunktion ist eine weitere einfache Aktivierungsfunktion. Neuronale Netzwerke wurden ursprünglich Perceptronen genannt. McCulloch & Pitts (1943) führten das ursprüngliche Perceptron ein und verwendeten eine Schrittaktivierungsfunktion wie die folgende Gleichung:[[Cite:mcculloch1943logical]](https://link.springer.com/article/10.1007/BF02478259) Die Schrittaktivierung ist 1, wenn x>=0,5, und andernfalls 0.

Diese Gleichung gibt einen Wert von 1,0 für eingehende Werte von 0,5 oder höher und 0 für alle anderen Werte aus. Schrittfunktionen, auch als Schwellenwertfunktionen bekannt, geben nur für Werte über dem angegebenen Schwellenwert 1 (wahr) zurück, wie in Abbildung 3.STEP zu sehen ist.

## Warum ReLU?

Warum ist die ReLU-Aktivierungsfunktion so beliebt? Eine der entscheidenden Verbesserungen an neuronalen Netzwerken macht Deep Learning möglich. [[Cite:nair2010rectified]](https://www.cs.toronto.edu/~hinton/absps/reluICML.pdf) Vor Deep Learning war die Sigmoid-Aktivierungsfunktion weit verbreitet. Wir haben die Sigmoid-Aktivierungsfunktion bereits früher in diesem Modul behandelt. Frameworks wie PyTorch trainieren neuronale Netzwerke häufig mit Gradientenabstieg. Damit das neuronale Netzwerk Gradientenabstieg verwenden kann, muss die Ableitung der Aktivierungsfunktion genommen werden. Das Programm muss partielle Ableitungen jedes der Gewichte für die Fehlerfunktion ableiten. Abbildung 3.DERV zeigt eine Ableitung, die momentane Änderungsrate.

**Abbildung 3.DERV: Ableitung**
![Derivative](https://data.heatonresearch.com/images/wustl/class/class_2_deriv.png "Derivative")

Die Ableitung der Sigmoidfunktion ist hier angegeben:

$$ \phi'(x)=\phi(x)(1-\phi(x)) $$

Lehrbücher geben diese Ableitung oft in anderen Formen an. Wir verwenden die obige Form aus Gründen der Recheneffizienz. Um zu sehen, wie wir diese Ableitung bestimmt haben, [refer to the following article](http://www.heatonresearch.com/aifh/vol3/deriv_sigmoid.html).

Wir stellen die Grafik der Sigmoid-Ableitung in Abbildung 3.SDERV dar.

**Abbildung 3.SDERV: Sigmoid-Ableitung**
![Sigmoid Derivative](https://data.heatonresearch.com/images/wustl/class/class_2_deriv_sigmoid.png "Sigmoid Derivative")

Die Ableitung sättigt sich schnell auf Null, wenn sich $x$ von Null wegbewegt. Dies ist kein Problem für die Ableitung der ReLU, die hier angegeben ist:

$$ \phi'(x) = \begin{cases} 1 & x > 0 \\ 0 & x \leq 0 \end{cases} $$
