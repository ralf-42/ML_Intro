# Teil 2.3: Gruppieren, Sortieren und Mischen

Wir werden uns einige Möglichkeiten ansehen, wie wir einen gesamten Pandas-Datenrahmen beeinflussen können. Mit diesen Techniken können wir Datensätze gruppieren, sortieren und mischen. Dies sind alles wichtige Vorgänge sowohl für die Vorverarbeitung als auch für die Auswertung von Daten.

## Einen Datensatz neu mischen

Möglicherweise verbergen sich in der Reihenfolge der Zeilen Ihres Datensatzes Informationen. Sofern Sie nicht mit Zeitreihendaten arbeiten, sollte die Reihenfolge der Zeilen nicht von Bedeutung sein. Überlegen Sie, ob Ihr Trainingssatz Mitarbeiter eines Unternehmens enthält. Vielleicht ist dieser Datensatz nach der Anzahl der Jahre sortiert, die die Mitarbeiter im Unternehmen beschäftigt waren. Es ist in Ordnung, eine einzelne Spalte zu haben, die die Dienstjahre angibt. Die Daten in dieser Reihenfolge anzuordnen, könnte jedoch problematisch sein.

Überlegen Sie, ob Sie die Daten in Training und Validierung aufteilen würden. Am Ende könnte Ihr Validierungssatz nur die neueren Mitarbeiter und der Trainingssatz die längerfristig beschäftigten Mitarbeiter enthalten. Das Aufteilen der Daten in eine k-fache Kreuzvalidierung könnte ähnliche Probleme mit sich bringen. Aus diesem Grund ist es wichtig, den Datensatz neu zu mischen.

Oft werden das Mischen und die Neuindizierung zusammen durchgeführt. Das Mischen randomisiert die Reihenfolge des Datensatzes. Die Zeilennummern von Pandas werden dadurch jedoch nicht geändert. Der folgende Code demonstriert eine Neumischung. Beachten Sie, dass das Programm die erste Spalte der Zeilenindizes nicht zurückgesetzt hat. Im Allgemeinen verursacht dies keine Probleme und ermöglicht die Rückverfolgung zur ursprünglichen Reihenfolge der Daten. Normalerweise ziehe ich es jedoch vor, diesen Index zurückzusetzen. Ich begründe dies damit, dass mir die Anfangsposition normalerweise egal ist und es einige Fälle gibt, in denen dieser ungeordnete Index Probleme verursachen kann.


```python
import numpy as np
import pandas as pd

df = read_csv(
    "https://data.heatonresearch.com/data/t81-558/auto-mpg.csv", na_values=["NA", "?"]
)

# np.random.seed(42) # Entfernen Sie das Kommentarzeichen aus dieser Zeile, um jedes Mal die gleiche Reihenfolge zu erhalten.
df = df.reindex(np.random.permutation(df.index))

set_option("display.max_columns", 7)
set_option("display.max_rows", 5)
display(df)
```

Der folgende Code demonstriert eine Neuindizierung. Beachten Sie, wie die Neuindizierung die Zeilenindizes sortiert.


```python
set_option("display.max_columns", 7)
set_option("display.max_rows", 5)

df.reset_index(inplace=True, drop=True)
display(df)
```

### Sortieren eines Datensatzes

Während es immer gut ist, einen Datensatz vor dem Training neu zu mischen, möchten Sie den Datensatz während des Trainings und der Vorverarbeitung möglicherweise auch sortieren. Durch das Sortieren des Datensatzes können Sie die Zeilen für eine oder mehrere Spalten entweder aufsteigend oder absteigend sortieren. Der folgende Code sortiert den MPG-Datensatz nach Namen und zeigt das erste Auto an.


```python
df = read_csv(
    "https://data.heatonresearch.com/data/t81-558/auto-mpg.csv", na_values=["NA", "?"]
)

df = df.sort_values(by="name", ascending=True)
print(f"Das erste Auto ist: {df['name'].iloc[0]}")

set_option("display.max_columns", 7)
set_option("display.max_rows", 5)
display(df)
```

### Gruppieren eines Datensatzes

Gruppieren ist eine typische Operation für Datensätze. In der Structured Query Language (SQL) wird diese Operation als „GROUP BY“ bezeichnet. Programmierer verwenden Gruppieren, um Daten zusammenzufassen. Dadurch verringert sich die Anzahl der zusammengefassten Zeilen normalerweise, und Sie können die Gruppierung nicht rückgängig machen. Aufgrund dieses Informationsverlusts ist es wichtig, die Originaldaten vor der Gruppierung beizubehalten.

Wir verwenden den Auto MPG-Datensatz, um die Gruppierung zu demonstrieren.


```python
df = read_csv(
    "https://data.heatonresearch.com/data/t81-558/auto-mpg.csv", na_values=["NA", "?"]
)

set_option("display.max_columns", 7)
set_option("display.max_rows", 5)
display(df)
```

Sie können den obigen Datensatz mit der Gruppe verwenden, um Zusammenfassungen durchzuführen. Der folgende Code gruppiert beispielsweise Zylinder nach dem Durchschnitt (Mittelwert). Dieser Code stellt die Gruppierung bereit. Zusätzlich zu **mean** können Sie andere Aggregatfunktionen verwenden, z. B. **sum** oder **count**.


```python
g = df.groupby("cylinders")["mpg"].mean()
g
```

Es könnte nützlich sein, diese **Durchschnittswerte** als Wörterbuch zu haben.


```python
d = g.to_dict()
d
```

Mit einem Wörterbuch können Sie schnell auf ein einzelnes Element zugreifen. Sie können beispielsweise schnell den Mittelwert für Sechszylinderautos nachschlagen. Sie werden sehen, dass die ZielKodierung, die später in diesem Modul vorgestellt wird, diese Technik verwendet.


```python
d[6]
```

Der folgende Code zeigt, wie die Anzahl der Zeilen gezählt wird, die der jeweiligen Zylinderanzahl entsprechen.


```python
df.groupby("cylinders")["mpg"].count().to_dict()
```
