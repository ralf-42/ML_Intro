# Teil 1.5: Funktionen, Lambdas und Map/Reduce

Mit Funktionen, **Lambdas** und **Map/Reduce** können Sie Ihre Daten auf fortgeschrittene Weise verarbeiten. Wir werden diese Techniken hier vorstellen und im nächsten Modul, in dem wir uns mit Pandas befassen, weiter darauf eingehen.

Funktionsparameter können in Python benannt oder unbenannt sein. Es können auch Standardwerte verwendet werden. Betrachten Sie die folgende Funktion.


```python
def say_hello(speaker, person_to_greet, greeting="Hello"):
print(f"{Begrüßung} {zu begrüßende Person}, hier ist {Speaker}.")


say_hello("Jeff", "John")
say_hello("Jeff", "John", "Goodbye")
say_hello(speaker="Jeff", person_to_greet="John", greeting="Goodbye")
```

Eine Funktion ist eine Möglichkeit, Code zu erfassen, der häufig ausgeführt wird. Betrachten Sie die folgende Funktion, mit der Leerzeichen aus einer Zeichenfolge entfernt werden können, sodass der erste Buchstabe groß geschrieben wird.


```python
def process_string(str):
    t = str.strip()
    return t[0].upper() + t[1:]
```

Diese Funktion lässt sich nun ganz einfach aufrufen.


```python
str = process_string("  hello  ")
print(f""{str}"")
```

Pythons **map** ist eine sehr nützliche Funktion, die in vielen verschiedenen Programmiersprachen zur Verfügung steht. Die **map**-Funktion nimmt eine **Liste** und wendet eine Funktion auf jedes Mitglied der **Liste** an und gibt eine zweite **Liste** zurück, die dieselbe Größe wie die erste hat.


```python
l = ["   apple  ", "pear ", "orange", "pine apple  "]
list(map(process_string, l))
```

## Karte

Die **map**-Funktion ist der Python-**Comprehension**, die wir zuvor untersucht haben, sehr ähnlich. Die folgende **Comprehension** führt dieselbe Aufgabe aus wie der vorherige Aufruf von **map**.


```python
l = ["   apple  ", "pear ", "orange", "pine apple  "]
l2 = [process_string(x) for x in l]
print(l2)
```

Die Entscheidung, eine **Map**-Funktion oder **Comprehension** zu verwenden, liegt beim Programmierer. Ich bevorzuge tendenziell **Map**, da es in anderen Programmiersprachen so üblich ist.


## Filter

Während eine **Map-Funktion** immer eine neue **Liste** mit derselben Größe wie das Original erstellt, erstellt die **Filter**-Funktion eine potenziell kleinere **Liste**.


```python
def greater_than_five(x):
    return x > 5


l = [1, 10, 20, 3, -2, 0]
l2 = list(filter(greater_than_five, l))
print(l2)
```

## Lambda

Es mag etwas mühsam erscheinen, eine ganze Funktion erstellen zu müssen, nur um zu prüfen, ob ein Wert größer als 5 ist. Ein **Lambda** erspart Ihnen diesen Aufwand. Ein Lambda ist im Wesentlichen eine unbenannte Funktion.


```python
l = [1, 10, 20, 3, -2, 0]
l2 = list(filter(lambda x: x > 5, l))
print(l2)
```

## Reduzieren

Zum Schluss werden wir **Reduce** verwenden. Wie **Filter** und **Map** funktioniert auch die **Reduce**-Funktion auf einer **Liste**. Das Ergebnis von **Reduce** ist jedoch ein einzelner Wert. Stellen Sie sich vor, Sie möchten die **Werte** einer **Liste** summieren. Die Summe wird durch ein **Lambda** implementiert.


```python
from functools import reduce

l = [1, 10, 20, 3, -2, 0]
result = reduce(lambda x, y: x + y, l)
print(result)
```
