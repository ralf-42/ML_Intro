## Aktuelle Spezifikationen des Atari 2600

* CPU: 1,19 MHz MOS-Technologie 6507
* Audio- und Videoprozessor: Television Interface Adapter (TIA)
* Spielfeldauflösung: 40 x 192 Pixel (NTSC). Es wird ein 20-Pixel-Register verwendet, das von links nach rechts gespiegelt oder kopiert wird, um eine Breite von 40 Pixeln zu erreichen.
* Spieler-Sprites: 8 x 192 Pixel (NTSC). Spieler-, Ball- und Raketen-Sprites verwenden Pixel, die 1/4 der Breite der Spielfeldpixel entsprechen (sofern sie nicht gestreckt werden).

* 128 Farben (NTSC). 128 auf dem Bildschirm möglich. Maximal 4 pro Zeile: Hintergrund, Spielfeld, Sprite von Spieler0 und Sprite von Spieler1. Palettenwechsel zwischen Zeilen sind üblich. Palettenwechsel während der Zeile sind möglich, aber aufgrund von Ressourcenbeschränkungen nicht üblich.

Im Spiel steuert der Spieler einen Schläger am unteren Bildschirmrand, mit dem er einen Ball gegen eine Ziegelwand am oberen Rand schlagen kann. Ziel ist es, alle Ziegel zu zerstören, indem man sie mit dem Ball trifft, den der Spieler mit dem Schläger abwehrt. Je weiter der Spieler kommt, desto schneller bewegt sich der Ball, und manche Ziegel müssen möglicherweise mehrmals getroffen werden, um zu zerbrechen. Der Spieler verliert eine Runde, wenn der Ball den Schläger verfehlt und den unteren Bildschirmrand trifft. Die Einfachheit des Gameplays von Breakout, kombiniert mit seinem steigenden Schwierigkeitsgrad im Spielverlauf, hat es zu einem Paradebeispiel für das leicht zu erlernende, aber schwer zu meisternde Designethos gemacht, das viele frühe Videospiele kennzeichnete.

In diesem Setup beobachtet der Agent den Spielstatus (normalerweise die Pixeldaten vom Bildschirm), wählt Aktionen aus (wie das Bewegen des Paddels nach links oder rechts) und erhält Belohnungen (wie die Punktzahl für das Zerbrechen von Steinen). ## Schulung des Agenten

Wir sind nun bereit, den DQN zu trainieren. Je nachdem, wie viele Episoden Sie durchlaufen möchten, kann dieser Vorgang mehrere Stunden dauern. Dieser Code aktualisiert sowohl den Verlust als auch den durchschnittlichen Ertrag während des Trainings. Mit zunehmendem Erfolg des Trainings sollte der durchschnittliche Ertrag steigen. Die gemeldeten Verluste spiegeln den durchschnittlichen Verlust für einzelne Trainingschargen wider.

```python
import gymnasium as gym
from stable_baselines3 import PPO
from stable_baselines3.common.evaluation import evaluate_policy
from stable_baselines3.common.env_util import make_atari_env
from stable_baselines3.common.vec_env import VecFrameStack

# Setzen Sie diese Konstante entweder auf 'Breakout' oder 'Atlantis', um das Spiel auszuwählen
GAME_NAME = "Breakout"  # Oder „Atlantis“

# Erstellen Sie die Spielumgebung. Beachten Sie, dass wir sie zur Vorverarbeitung mit VecFrameStack umschließen
env_id = f"{GAME_NAME}NoFrameskip-v4"
env = make_atari_env(env_id, n_envs=4, seed=0)
env = VecFrameStack(env, n_stack=4)

# Initialisieren Sie den Agenten. Hier verwenden wir Proximal Policy Optimization (PPO).
model = PPO("CnnPolicy", env, verbose=1, tensorboard_log="./atari_ppo_tensorboard/")

# Trainieren Sie den Agenten
TIMESTEPS = 1e5
model.learn(total_timesteps=TIMESTEPS)

# Speichern des Modells
model.save(f"{GAME_NAME}_ppo_model")

# Bewerten Sie den geschulten Agenten
mean_reward, std_reward = evaluate_policy(model, env, n_eval_episodes=10)

print(f"Mittlere Belohnung: {mean_reward} +/- {std_reward}")

# Vergessen Sie nicht, die Umgebung zu schließen, wenn Sie fertig sind
env.close()
```

## Videos

Die vielleicht überzeugendste Möglichkeit, die Ergebnisse eines Atari-Spiels anzuzeigen, ist ein Video, das es uns ermöglicht, dem Agenten beim Spielen zuzusehen. Wir haben jetzt ein trainiertes Modell und können seinen Trainingsfortschritt in einem Diagramm beobachten. Die folgenden Funktionen sind definiert, um dem Agenten beim Spielen des Spiels im Notebook zuzusehen.

```python
from stable_baselines3.common.vec_env import VecFrameStack
from stable_baselines3.common.env_util import make_atari_env
from stable_baselines3.common.vec_env import VecVideoRecorder
from stable_baselines3 import PPO
import os

# Legen Sie hier den Spielnamen fest
GAME_NAME = "Breakout"  # Kann auch 'Atlantis' sein

# Laden Sie Ihr zuvor trainiertes Modell
model_path = f"{GAME_NAME}_ppo_model.zip"
model = PPO.load(model_path)

# Erstellen Sie die Atari-Umgebung und wenden Sie die richtigen Wrapper an
env_id = f"{GAME_NAME}NoFrameskip-v4"
env = make_atari_env(env_id, n_envs=1, seed=0)
env = VecFrameStack(env, n_stack=4)

# Die Umgebung aufzeichnen
video_folder = "/content/videos"
if not os.path.exists(video_folder):
    os.makedirs(video_folder)

env = VecVideoRecorder(
    env,
    video_folder,
    record_video_trigger=lambda step: step == 0,
    video_length=500,
    name_prefix=f"{GAME_NAME}-agent",
)

# Setzen Sie die Umgebung zurück und beobachten Sie die anfängliche Beobachtungsform
obs = env.reset()
print("Initial observation shape:", obs.shape)  # Sollte (1, 4, 84, 84) sein

# Führen Sie eine Episode aus
done = False
while not done:
    action, _states = model.predict(obs, deterministic=True)
    obs, rewards, done, info = env.step(action)
    env.render()

# Schließen Sie die Umgebung, wodurch auch das Video gespeichert werden soll.
env.close()
```
