## Weitere Beispiele: Umgang mit Adressen

Das Kodieren von Adressen in ein neuronales Netzwerk kann schwierig sein. Es gibt viele verschiedene Ansätze, und Sie müssen überlegen, wie Sie die Adresse in etwas Aussagekräftigeres umwandeln können. Kartenkoordinaten können ein guter Ansatz sein. [latitude and longitude](https://en.wikipedia.org/wiki/Geographic_coordinate_system) kann eine nützliche Kodierung sein. Dank der Leistungsfähigkeit des Internets ist es relativ einfach, eine Adresse in ihre Breiten- und Längengrade umzuwandeln. Der folgende Code bestimmt die Koordinaten von [latitude and longitude](https://en.wikipedia.org/wiki/Geographic_coordinate_system):

```python
import requests

address = "1 Brookings Dr, St. Louis, MO 63130"

response = requests.get(
    "https://maps.googleapis.com/maps/api/geocode/json?key={}&address={}".format(
        GOOGLE_KEY, address
    )
)

resp_json_payload = response.json()

if "error_message" in resp_json_payload:
    print(resp_json_payload["error_message"])
else:
    print(resp_json_payload["results"][0]["geometry"]["location"])
```

Sie sind möglicherweise nicht besonders hilfreich, wenn Sie Breiten- und Längengrad als zwei Merkmale in das neuronale Netzwerk eingeben. Mit diesen beiden Werten könnte Ihr neuronales Netzwerk Standorte auf einer Karte gruppieren. Manchmal kann es nützlich sein, Standorte auf einer Karte zu gruppieren. Allerdings werden Sie die Koordinaten oft in Entfernungen umwandeln wollen. Es ist relativ einfach, die Entfernung zwischen zwei beliebigen Punkten auf der Erde abzuschätzen, indem Sie den [great circle distance](https://en.wikipedia.org/wiki/Great-circle_distance) zwischen zwei beliebigen Punkten auf einer Kugel verwenden:

Der folgende Code implementiert diese Formel:

$$ \Delta\sigma=\arccos\bigl(\sin\phi_1\cdot\sin\phi_2+\cos\phi_1\cdot\cos\phi_2\cdot\cos(\Delta\lambda)\bigr) $$

$$ d = r \, \Delta\sigma $$

```python
from math import sin, cos, sqrt, atan2, radians

URL = "https://maps.googleapis.com" + "/maps/api/geocode/json?key={}&address={}"


# Distanzfunktion
def distance_lat_lng(lat1, lng1, lat2, lng2):
    # ungefährer Radius der Erde in km
    R = 6373.0

    # Grad in Radiant (Breite/Länge sind in Grad angegeben)
    lat1 = radians(lat1)
    lng1 = radians(lng1)
    lat2 = radians(lat2)
    lng2 = radians(lng2)

    dlng = lng2 - lng1
    dlat = lat2 - lat1

    a = sin(dlat / 2) ** 2 + cos(lat1) * cos(lat2) * sin(dlng / 2) ** 2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    return R * c


# Suchen Sie nach Breitengraden und Längengraden für die Adresse
def lookup_lat_lng(address):
    response = requests.get(URL.format(GOOGLE_KEY, address))
    json = response.json()
    if len(json["results"]) == 0:
        raise ValueError("Google API error on: {}".format(address))
    map = json["results"][0]["geometry"]["location"]
    return map["lat"], map["lng"]


# Entfernung zwischen zwei Standorten

import requests

address1 = "1 Brookings Dr, St. Louis, MO 63130"
address2 = "3301 College Ave, Fort Lauderdale, FL 33314"

lat1, lng1 = lookup_lat_lng(address1)
lat2, lng2 = lookup_lat_lng(address2)

print(
    "Distance, St. Louis, MO to Ft. Lauderdale, FL: {} km".format(
        distance_lat_lng(lat1, lng1, lat2, lng2)
    )
)
```

Entfernungen können ein nützliches Mittel zum Kodieren von Adressen sein. Es wäre hilfreich, wenn Sie überlegen würden, welche Entfernung für Ihren Datensatz hilfreich sein könnte. Bedenken Sie:

* Entfernung zu einer Großstadt
* Entfernung zu einem Konkurrenten
* Entfernung zu einem Distributionszentrum
* Entfernung zu einer Verkaufsstelle

Der folgende Code berechnet die Entfernung zwischen 10 Universitäten und der Washington University in St. Louis:

```python
# Kodierung anderer Universitäten nach ihrer Entfernung zur Washington University

schools = [
    ["Princeton University, Princeton, NJ 08544", "Princeton"],
    ["Massachusetts Hall, Cambridge, MA 02138", "Harvard"],
    ["5801 S Ellis Ave, Chicago, IL 60637", "University of Chicago"],
    ["Yale, New Haven, CT 06520", "Yale"],
    ["116th St & Broadway, New York, NY 10027", "Columbia University"],
    ["450 Serra Mall, Stanford, CA 94305", "Stanford"],
    ["77 Massachusetts Ave, Cambridge, MA 02139", "MIT"],
    ["Duke University, Durham, NC 27708", "Duke University"],
    [
        "University of Pennsylvania, Philadelphia, PA 19104",
        "University of Pennsylvania",
    ],
    ["Johns Hopkins University, Baltimore, MD 21218", "Johns Hopkins"],
]

lat1, lng1 = lookup_lat_lng("1 Brookings Dr, St. Louis, MO 63130")

for address, name in schools:
    lat2, lng2 = lookup_lat_lng(address)
    dist = distance_lat_lng(lat1, lng1, lat2, lng2)
    print("School '{}', distance to wustl is: {}".format(name, dist))
```
