# Teil 3.5: Sequenzen vs. Klassen in PyTorch

Zuvor haben wir die Grundlagen von PyTorch erkundet und erfahren, wie man neuronale Netzwerke mithilfe des Moduls **nn.Sequential** definiert. Mit diesem Ansatz konnten wir neuronale Netzwerke erstellen, indem wir eine Abfolge von Schichten unkompliziert und präzise spezifizierten. Obwohl **nn.Sequential** die primäre Methode ist, die wir in diesem Kurs verwenden werden, ist es wichtig, eine alternative Möglichkeit zu verstehen, neuronale Netzwerke als Klasse zu definieren.

## Neuronale Netze als Sequenzen definieren

Bevor wir den klassenbasierten Ansatz überprüfen, fassen wir noch einmal zusammen, wie wir neuronale Netzwerke mithilfe des Moduls **nn.Sequential** definieren. In PyTorch stellen Sie ein neuronales Netzwerk als eine Abfolge von Schichten dar. Jede Schicht führt eine bestimmte Berechnung der Eingabedaten durch und übergibt die transformierte Ausgabe an die nächste Schicht.

Hier ist ein kurzes Beispiel für die Definition eines einfachen neuronalen Netzwerks mit **nn.Sequential**:

```
import torch
import torch.nn as nn

# Define the neural network architecture
model = nn.Sequential(
    nn.Linear(784, 128),
    nn.ReLU(),
    nn.Linear(128, 64),
    nn.ReLU(),
    nn.Linear(64, 10),
    nn.LogSoftmax(dim=1)
)

```

In diesem Beispiel haben wir ein neuronales Netzwerk mit drei linearen Schichten erstellt, die mit ReLU-Aktivierungsfunktionen durchsetzt sind. Die letzte Schicht verwendet die **LogSoftmax**-Aktivierung, die wir häufig für Klassifizierungsaufgaben verwenden.

## Neuronale Netze als Klassen definieren

Während der nn.Sequential-Ansatz in vielen Fällen intuitiv und praktisch ist, kann er einschränkend wirken, wenn wir mehr Flexibilität in der Netzwerkarchitektur benötigen. Durch die Definition neuronaler Netzwerke als Klassen können wir benutzerdefinierte Architekturen mit komplexen Verhaltensweisen und gemeinsam nutzbaren Komponenten erstellen.

Um ein neuronales Netzwerk als Klasse zu definieren, gliedern wir normalerweise die von PyTorch bereitgestellte Klasse **nn.Module** in eine Unterklasse. Diese Basisklasse bietet grundlegende Funktionen zum Organisieren der Netzwerkparameter und zum Handhaben von Berechnungen während Vorwärts- und Rückwärtsdurchläufen.

Lassen Sie uns dies veranschaulichen, indem wir unser vorheriges Beispiel mit einem klassenbasierten Ansatz umschreiben:


```
import torch
import torch.nn as nn

class CustomNetwork(nn.Module):
    def __init__(self):
        super(CustomNetwork, self).__init__()
        self.fc1 = nn.Linear(784, 128)
        self.relu1 = nn.ReLU()
        self.fc2 = nn.Linear(128, 64)
        self.relu2 = nn.ReLU()
        self.fc3 = nn.Linear(64, 10)
        self.log_softmax = nn.LogSoftmax(dim=1)

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu1(x)
        x = self.fc2(x)
        x = self.relu2(x)
        x = self.fc3(x)
        x = self.log_softmax(x)
        return x

# Create an instance of the CustomNetwork
model = CustomNetwork()

```

In diesem Beispiel definieren wir eine Klasse namens CustomNetwork, die von nn.Module erbt. Innerhalb der Klasse beschreiben wir die Schichten des Netzwerks als Attribute. Die Vorwärtsmethode gibt an, wie die Eingabe während des Vorwärtsdurchlaufs durch diese Schichten fließt.

Indem wir neuronale Netzwerke als Klassen definieren, können wir komplexere Architekturen erstellen, bedingte Logik innerhalb des Netzwerks nutzen und wiederverwendbare Komponenten kapseln. Diese Flexibilität wird besonders nützlich, wenn wir uns im Laufe des Kurses mit fortgeschrittenen Themen befassen.

## Die richtige Methode wählen

Während sowohl der nn.Sequential- als auch der klassenbasierte Ansatz es uns ermöglichen, neuronale Netzwerke zu definieren, werden wir in diesem Kurs aufgrund seiner Einfachheit und Benutzerfreundlichkeit hauptsächlich nn.Sequential verwenden. Es bietet eine effiziente Möglichkeit, Netzwerke mit einer linearen Abfolge von Schichten zu konstruieren.

Das Verständnis des klassenbasierten Ansatzes ist jedoch entscheidend für eine tiefere Anpassung und den Aufbau komplexerer Architekturen. Wenn Sie PyTorch immer besser beherrschen, werden Sie auf Szenarien stoßen, in denen es notwendig wird, Netzwerke als Klassen zu definieren. Wenn Sie beide Methoden beherrschen, sind Sie gut gerüstet, um eine breite Palette von Herausforderungen beim Entwurf neuronaler Netzwerke zu bewältigen.
