# Teil 7.1: Einführung in Generative KI

Generative KI ist ein faszinierendes und sich rasch entwickelndes Teilgebiet der künstlichen Intelligenz, das sich auf die Schaffung neuer Inhalte konzentriert, von Kunst, Musik und Geschichten bis hin zu eher technischen Anwendungen wie der Generierung von Code oder der Vorhersage komplexer Systeme. Dies steht im Gegensatz zu anderen Arten von KI, die sich mehr mit der Analyse und Interpretation vorhandener Daten befassen, wie etwa der Klassifizierung von Bildern oder der Vorhersage von Aktienkursen. Während herkömmliche KI-Systeme so konzipiert sind, dass sie aus einer Reihe von Möglichkeiten die genaueste oder optimale Antwort geben, sind generative KI-Systeme darauf ausgelegt, neue Möglichkeiten zu schaffen und neue Daten zu generieren, die in gewisser Weise den Daten ähneln, mit denen sie trainiert wurden.

Ein generatives KI-System könnte beispielsweise anhand eines großen Datensatzes klassischer Musik trainiert werden und dann in der Lage sein, neue Kompositionen im Stil von Bach oder Beethoven zu erstellen. Eine weitere Anwendung könnte darin bestehen, das Modell anhand einer großen Menge an Textdaten zu trainieren, um einen Chatbot zu erstellen, der menschenähnliche Antworten auf Fragen oder Eingabeaufforderungen generieren kann, wie wir es bei den GPT-Modellen von OpenAI gesehen haben. Weitere Beispiele sind die Generierung realistischer Bilder, die Erstellung virtueller Umgebungen oder sogar die Entwicklung von Molekülen für neue Medikamente.

Diese bemerkenswerte Fähigkeit der generativen KI eröffnet eine Welt voller Möglichkeiten und Anwendungen, die gerade erst erforscht werden. In diesem Abschnitt werden wir tiefer in die Funktionsweise der generativen KI eintauchen, die Technologien untersuchen, die sie antreiben, wie Generative Adversarial Networks (GANs) und Large Language Models (LLMs), und einige der aufregendsten Anwendungen und zukünftigen Richtungen dieses innovativen Feldes erkunden.

## Die Anfänge der generativen KI

Die Geschichte der generativen KI lässt sich bis in die Anfänge der künstlichen Intelligenz selbst zurückverfolgen. In den 1950er und 60er Jahren experimentierten Forscher bereits mit Algorithmen, die neue Inhalte generieren konnten, wie zum Beispiel der 1966 am MIT entwickelte Chatbot ELIZA, der auf Benutzereingaben Antworten in einer Weise generieren konnte, die einem Psychotherapeuten ähnelte. Das wahre Potenzial der generativen KI blieb jedoch bis zum Aufkommen leistungsstärkerer Computerhardware und größerer Datensätze im 21. Jahrhundert weitgehend ungenutzt. Während ELIZA für die damalige Zeit fortschrittlich schien, fehlte seinen Gesprächen jede nennenswerte Tiefe:

```
> Hello, I am Eliza. 
* I am afraid I will never learn Python.
> Do you believe it is normal to be afraid you will never learn Python? 
* Yes
> I see.
```

Im Jahr 2006 entwickelten Geoffrey Hinton und seine Kollegen an der Universität von Toronto eine Art neuronales Netzwerk namens Deep Belief Network (DBN), das darauf trainiert werden konnte, neue Daten zu generieren, die den Daten ähnelten, mit denen es trainiert wurde. Dies war ein wichtiger Meilenstein in der Entwicklung der generativen KI, da es zeigte, dass neuronale Netzwerke nicht nur zum Klassifizieren und Analysieren von Daten, sondern auch zum Erstellen neuer Daten verwendet werden konnten. Anfangs wurde dieses DBN darauf trainiert, MNIST-Ziffern zu erkennen, wie in Abbildung 7.MNIST zu sehen ist.

Der nächste große Durchbruch kam 2014 mit der Einführung von Generative Adversarial Networks (GANs) durch Ian Goodfellow und seine Kollegen. GANs bestehen aus zwei neuronalen Netzwerken, einem Generator und einem Diskriminator, die gleichzeitig durch eine Art Spiel trainiert werden. Der Generator versucht, Daten zu erstellen, die von echten Daten nicht zu unterscheiden sind, während der Diskriminator versucht, zu erkennen, ob die Daten echt oder gefälscht sind. Dieser kontroverse Trainingsprozess führt dazu, dass der Generator immer besser darin wird, realistische Daten zu erstellen, und wurde verwendet, um alles zu generieren, von realistischen Bildern nicht existierender Gesichter bis hin zu neuen Videospiel-Levels.

Stabile Diffusion ist eine neue Entwicklung im Bereich der generativen KI, die große Aufmerksamkeit erregt hat. Traditionelle Methoden wie GANs und VAEs haben ihre Nachteile, wie z. B. den Moduskollaps in GANs. Stabile Diffusion hingegen verwendet einen anderen Ansatz, indem ein diffusionsbasiertes Wahrscheinlichkeitsmodell übernommen wird. Bei diesem Ansatz werden die Daten so transformiert, dass sie sich ausbreiten oder „diffundieren“, und der Prozess wird dann umgekehrt ausgeführt, um neue Daten zu generieren. Das Modell wird trainiert, indem der nächste Diffusionsschritt im umgekehrten Prozess aus einem bestimmten Zustand vorhergesagt wird, wodurch es eine detaillierte und qualitativ hochwertige Darstellung der Daten erlernen kann. Dies führt im Vergleich zu anderen Methoden zur Generierung realistischerer und detaillierterer Bilder. Darüber hinaus haben sich stabile Diffusionsmodelle im Vergleich zu GANs als stabiler und einfacher zu trainieren erwiesen, was sie zu einer vielversprechenden Alternative für die Generierung hochwertiger Inhalte in verschiedenen Anwendungen macht.

Mit stabiler Diffusion können Sie Eingabeaufforderungen zum Rendern des Bildes verwenden, was Ihnen eine große Kontrolle gibt. Abbildung 7.GAN-STYLE zeigt dieselbe Frau in mehreren Posen und Einstellungen.

Diese Modelle werden anhand riesiger Mengen von Textdaten trainiert und können menschenähnliche Texte zu einer Vielzahl von Themen generieren.
