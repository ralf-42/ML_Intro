Wie die meisten modernen Programmiersprachen enthält Python Listen, Sets, Wörterbücher und andere Datenstrukturen als integrierte Typen. Die Syntax beider ähnelt JSON. Es ist wichtig, die Unterschiede zwischen diesen drei grundlegenden Sammlungstypen zu verstehen.

- **Wörterbuch** – Ein Wörterbuch ist eine veränderbare, ungeordnete Sammlung, die Python mit Namens- und Wertepaaren indiziert.
- **Liste** – Eine Liste ist eine veränderbare, geordnete Sammlung, die doppelte Elemente zulässt.
- **Set** – Ein Set ist eine veränderbare, ungeordnete Sammlung ohne doppelte Elemente.
- **Tupel** – Ein Tupel ist eine unveränderliche geordnete Sammlung, die doppelte Elemente zulässt.

Die meisten Python-Sammlungen sind veränderbar, d. h. das Programm kann nach der Definition Elemente hinzufügen und entfernen. Eine unveränderliche Sammlung kann nach der Definition keine Elemente hinzufügen oder entfernen. Es ist auch wichtig zu verstehen, dass eine geordnete Sammlung bedeutet, dass die Elemente ihre Reihenfolge beibehalten, während das Programm sie einer Sammlung hinzufügt. Diese Reihenfolge muss keine bestimmte Reihenfolge sein, z. B. alphabetisch oder numerisch.

Listen und Tupel sind in Python sehr ähnlich und werden oft verwechselt. Der wesentliche Unterschied besteht darin, dass eine Liste veränderbar ist, ein Tupel jedoch nicht. Wir schließen also eine Liste ein, wenn wir ähnliche Elemente enthalten möchten, und ein Tupel, wenn wir im Voraus wissen, welche Informationen darin enthalten sein sollen.

Viele Programmiersprachen enthalten eine Datensammlung namens Array. Der Array-Typ fehlt in Python merklich. Im Allgemeinen verwendet der Programmierer in Python eine Liste anstelle eines Arrays. Diese Einschränkung führt zu den berüchtigten Array-Overrun-Bugs und Sicherheitsproblemen. Die Python-Liste ist viel flexibler, da das Programm die Größe einer Liste dynamisch ändern kann.

### Listen und Tupel

Für ein Python-Programm sind Listen und Tupel sehr ähnlich. Sowohl Listen als auch Tupel enthalten eine geordnete Sammlung von Elementen. Als Programmierer kann man damit auskommen, nur Listen zu verwenden und Tupel zu ignorieren.

Der wichtigste syntaktische Unterschied besteht darin, dass eine Liste in eckige Klammern eingeschlossen ist: [], and a tuple is enclosed by parenthesis (). Der folgende Code definiert sowohl eine Liste als auch ein Tupel.


````python
l = ["a", "b", "c", "d"]
t = ("a", "b", "c", "d")

print(l)
print(t)
````

Der Hauptunterschied, den Sie programmgesteuert sehen werden, besteht darin, dass eine Liste veränderbar ist, was bedeutet, dass das Programm sie ändern kann. Ein Tupel ist unveränderlich, was bedeutet, dass das Programm es nicht ändern kann. Der folgende Code zeigt, dass das Programm eine Liste ändern kann. Dieser Code zeigt auch, dass Python Listen beginnend bei Element 0 indiziert. Der Zugriff auf Element eins ändert das zweite Element in der Sammlung. Ein Vorteil von Tupeln gegenüber Listen besteht darin, dass Tupel im Allgemeinen etwas schneller zu durchlaufen sind als Listen.


````python
l[1] = "changed"
# t[1] = 'changed' # Dies würde zu einem Fehler führen

print(l)
````

Wie viele andere Sprachen verfügt Python über eine For-Each-Anweisung. Mit dieser Anweisung können Sie jedes Element einer Sammlung, beispielsweise einer Liste oder eines Tupels, durchlaufen.


````python
# Über eine Sammlung iterieren.
for s in l:
    print(s)
````

Die Funktion **enumerate** ist nützlich, um eine Sammlung aufzuzählen und auf den Index des Elements zuzugreifen, bei dem wir uns gerade befinden.


````python
# Iterieren Sie über eine Sammlung und wissen Sie, wo Ihr Index ist. (Python ist nullbasiert!)
for i, l in enumerate(l):
print(f"{i}:{l}")
````

Einer **Liste** können mehrere Objekte hinzugefügt werden, z. B. Zeichenfolgen. Doppelte Werte sind zulässig. **Tupel** erlauben dem Programm nicht, nach der Definition zusätzliche Objekte hinzuzufügen.


````python
# Elemente manuell hinzufügen, Listen erlauben Duplikate
c = []
c.append("a")
c.append("b")
c.append("c")
c.append("c")
print(c)
````

Bei geordneten Sammlungen, wie Listen und Tupeln, können Sie auf ein Element über seine Indexnummer zugreifen, wie im folgenden Code. Bei ungeordneten Sammlungen, wie Wörterbüchern und Mengen, kann das Programm nicht auf diese Weise darauf zugreifen.


````python
print(c[1])
````

## Sätze

Ein Python-**Set** enthält eine ungeordnete Sammlung von Objekten, aber Sets erlauben _keine_ Duplikate. Wenn ein Programm einem Set ein doppeltes Element hinzufügt, verbleibt nur eine Kopie jedes Elements in der Sammlung. Das Hinzufügen eines doppelten Elements zu einem Set führt nicht zu einem Fehler. Jede der folgenden Techniken definiert ein Set.


````python
s = set()
s = {"a", "b", "c"}
s = set(["a", "b", "c"])
print(s)
````

Eine **Liste** ist immer in eckige Klammern eingeschlossen, [], a **tuple** in parenthesis (), und ebenso ist ein **Set** in geschweifte Klammern eingeschlossen. Programme können während der Ausführung Elemente zu einem **Set** hinzufügen. Programme können mit der Funktion **add** dynamisch Elemente zu einem **Set** hinzufügen. Es ist wichtig zu beachten, dass die Funktion **append** Elemente zu Listen hinzufügt, während die Funktion **add** Elemente zu einem **Set** hinzufügt.


````python
# Elemente manuell hinzufügen, Sets erlauben keine Duplikate
# Sets hinzufügen, Listen anhängen. Ich finde das nervig.
c = set()
c.add("a")
c.add("b")
c.add("c")
c.add("c")
print(c)
````

## Weitere erweiterte Listen

Für die in diesem Abschnitt vorgestellten Listen stehen mehrere erweiterte Funktionen zur Verfügung. Eine solche Funktion ist **zip**. Mit dem Befehl **zip** können zwei Listen zu einer einzigen Liste kombiniert werden. Der folgende Code demonstriert den Befehl **zip**.


````python
a = [1, 2, 3, 4, 5]
b = [5, 4, 3, 2, 1]

print(zip(a, b))
````

Um die Ergebnisse der **zip**-Funktion anzuzeigen, konvertieren wir das zurückgegebene Zip-Objekt in eine Liste. Wie Sie sehen, gibt die **zip**-Funktion eine Liste von Tupeln zurück. Jedes Tupel stellt ein Paar von Elementen dar, die die Funktion zusammengezippt hat. Die Reihenfolge in den beiden Listen wurde beibehalten.


````python
a = [1, 2, 3, 4, 5]
b = [5, 4, 3, 2, 1]

print(list(zip(a, b)))
````

Die übliche Methode zur Verwendung des Befehls **zip** ist innerhalb einer For-Schleife. Der folgende Code zeigt, wie eine For-Schleife jeder Sammlung, die das Programm durchläuft, eine Variable zuweisen kann.


````python
a = [1, 2, 3, 4, 5]
b = [5, 4, 3, 2, 1]

for x, y in zip(a, b):
print(f"{x} - {y}")
````

Normalerweise haben beide Sammlungen die gleiche Länge, wenn sie an den **zip**-Befehl übergeben werden. Es ist kein Fehler, wenn Sammlungen unterschiedlicher Länge vorhanden sind. Wie der folgende Code zeigt, verarbeitet der **zip**-Befehl nur Elemente bis zur Länge der kleineren Sammlung.


````python
a = [1, 2, 3, 4, 5]
b = [5, 4, 3]

print(list(zip(a, b)))
````

Manchmal möchten Sie vielleicht den aktuellen numerischen Index wissen, wenn eine For-Schleife eine geordnete Sammlung durchläuft. Verwenden Sie den Befehl **enumerate**, um den Indexspeicherort für ein Sammlungselement zu verfolgen. Da der Befehl **enumerate** mit numerischen Indizes der Sammlung arbeitet, weist der Befehl zip Elementen aus ungeordneten Sammlungen beliebige Indizes zu.

Überlegen Sie, wie Sie ein Python-Programm konstruieren könnten, um jedes Element, das größer als 5 ist, in den Wert 5 zu ändern. Das folgende Programm führt diese Transformation durch. Der Befehl enumerate ermöglicht es der Schleife, zu erkennen, bei welchem ​​Elementindex sie sich gerade befindet, sodass das Programm den Wert des aktuellen Elements der Sammlung ändern kann.


````python
a = [2, 10, 3, 11, 10, 3, 2, 1]
for i, x in enumerate(a):
    if x > 5:
        a[i] = 5
print(a)
````

Mit dem Comprehension-Befehl kann dynamisch eine Liste aufgebaut werden. Die folgende Comprehension zählt von 0 bis 9 und fügt jeden Wert (multipliziert mit 10) zu einer Liste hinzu.


````python
lst = [x * 10 for x in range(10)]
print(lst)
````
