# Teil 6.2: LangChain, ChatGPT und NLP

Large Language Models (LLMs) wie GPT haben KI in den Mainstream gebracht. LLMs ermöglichen es normalen Benutzern, mithilfe natürlicher Sprache mit KI zu interagieren. Die meisten dieser Sprachmodelle erfordern extreme Verarbeitungskapazitäten und Hardware. Aus diesem Grund werden über das Internet zugängliche Anwendungsprogrammierschnittstellen (APIs) zu gängigen Einstiegspunkten für diese Modelle. Eines der überzeugendsten Merkmale von Diensten wie ChatGPT ist ihre Verfügbarkeit als API. Doch bevor wir uns in die Tiefen der Kodierung und Integration stürzen, wollen wir verstehen, was eine API ist und welche Bedeutung sie im KI-Bereich hat.

API steht für Application Programming Interface (Anwendungsprogrammierschnittstelle). Stellen Sie es sich als eine Brücke oder einen Messenger vor, über den zwei verschiedene Softwareanwendungen miteinander kommunizieren können. Im Kontext von KI und maschinellem Lernen ermöglichen APIs Entwicklern häufig den Zugriff auf ein bestimmtes Modell oder einen bestimmten Dienst, ohne das Modell auf ihrem lokalen Computer unterbringen zu müssen. Diese Technik kann von Vorteil sein, wenn das betreffende Modell, wie ChatGPT, groß und ressourcenintensiv ist.

Im Bereich der künstlichen Intelligenz bieten APIs mehrere entscheidende Vorteile:

* Skalierbarkeit: Da das eigentliche Modell auf externen Servern läuft, müssen sich Entwickler keine Gedanken über die Skalierung der Infrastruktur machen.
* Wartung: Sie können die neueste und beste Version des Modells verwenden, ohne Ihre lokale Kopie ständig aktualisieren zu müssen.
* Kostengünstig: Die Nutzung externer Rechenressourcen kann kostengünstiger sein als die lokale Wartung einer High-End-Infrastruktur, insbesondere bei sporadischen oder einmaligen Aufgaben.
* Benutzerfreundlichkeit: Anstatt sich mit den kleinsten Details der Modellimplementierung und -optimierung befassen zu müssen, können Entwickler die Funktionen mit wenigen Codezeilen direkt nutzen.

In diesem Abschnitt führen wir die neuronalen Netzwerkberechnungen nicht lokal aus. Stattdessen kommuniziert unser PyTorch-Code mit der OpenAI-API, um auf die Fähigkeiten von ChatGPT zuzugreifen und diese zu nutzen. Die eigentliche Ausführung des neuronalen Netzwerkcodes erfolgt auf OpenAI-Servern, wodurch eine einzigartige Synergie aus der Flexibilität von PyTorch und der Konversationskompetenz von ChatGPT entsteht.

## Arbeiten mit Eingabeaufforderungen

Wir müssen häufig komplexe Eingabeaufforderungen erstellen, die mehrere Variablen in die endgültige Eingabeaufforderung integrieren. Dazu können wir die normale Python-String-Verarbeitung verwenden. Lassen Sie uns ChatGPT verwenden, um von Französisch nach Englisch zu übersetzen, und dabei normale Python-F-Strings verwenden, um die Eingabeaufforderung zu erstellen.

````python
text = """Laissez les bons temps rouler"""
style = "American English"

prompt = f"""Translate the text \
that is delimited by triple backticks \
into a style that is {style}. \
text: ```{text}```
"""

response = llm.invoke(prompt)

# Drucken Sie die Antwort
print(response.content)
````


Wir können LangChain zum Erstellen dynamischer Eingabeaufforderungen verwenden.

````python
from langchain.prompts import ChatPromptTemplate

template_text = """Translate the text \
that is delimited by triple backticks \
into a style that is {style}. \
text: ```{text}```
"""

prompt_template = ChatPromptTemplate.from_template(template_text)
````

Wir können jetzt die Lücken für diese Eingabeaufforderung ausfüllen und die erstellte Eingabeaufforderung beobachten, bei der es sich um eine Textzeichenfolge handelt.

````python
prompt = prompt_template.format_messages(
    style="American English", text="千里之行，始于足下。"
)

print(type(prompt))
print(type(prompt[0]))

print(prompt[0])
````

Diese neu erstellte Eingabeaufforderung kann nun die beabsichtigte Übersetzungsaufgabe ausführen.

````python
# Rufen Sie den LLM an, um den Stil der Kundennachricht zu übersetzen
response = llm.invoke(prompt)
print(response)
````

## Ausgabe verarbeiten

Wir werden uns nun ein komplexeres Beispiel für die Textextraktion ansehen und sehen, wie LangChain uns helfen kann, mehrere von ChatGPT zurückgegebene Werte zu extrahieren. Hier werden wir sehen, wie drei Felder aus einer Produktbeschreibung extrahiert werden können.

````python
from langchain.output_parsers import ResponseSchema
from langchain.output_parsers import StructuredOutputParser

material_schema = ResponseSchema(
    name="material",
    description="What is the material that this \
                             item is made of? If unknown, make an estimate.",
)
description_schema = ResponseSchema(
    name="shape",
    description="What is the shape of this \
                                      item? If unknown, return null.",
)
who_schema = ResponseSchema(
    name="who",
    description="Who is the likely user of \
                                    this item? If unkown, make an estimate.",
)

response_schemas = [material_schema, description_schema, who_schema]
````

Wie Sie dem obigen Code entnehmen können, extrahieren wir drei Felder aus der Produktbeschreibung: Material, Form und Wer. Wir beschreiben LangChain für jedes dieser Felder, um LangChain mitzuteilen, was jedes Feld ist, was dabei hilft, es in der Produktbeschreibung zu finden. Als Nächstes konstruieren wir einen **StructuredOutputParser**, um diese Daten tatsächlich abzurufen.

````python
output_parser = StructuredOutputParser.from_response_schemas(response_schemas)
format_instructions = output_parser.get_format_instructions()

prompt = ChatPromptTemplate.from_template(
    template="""
For the product text, extract this information in valid JSON, with commas. \
{format_instructions}
Text:
{text}"""
)

product_description = """\
Ross Brachiosaurus has a gentle spirit that any dog will quickly love. \
His long body lends itself for tossing or tugging alone or with friends, \
and his size makes him an excellent cuddle companion.
"""
````

Wir können nun diese drei Teile ausführen: den Prompt Builder, das LLM selbst aufrufen und dann schließlich die Ausgabe des LLM analysieren. Dadurch erhalten wir ein Wörterbuch, das diese drei Felder enthält.

````python
result1 = prompt.invoke(
    {"text": product_description, "format_instructions": format_instructions}
)
result2 = llm.invoke(result1)
result3 = output_parser.invoke(result2)
print(result3)
````

Oftmals verfügen Sie in langchain über mehrere Komponenten, die Sie in einer „Kette“ aufrufen müssen; hierzu können Sie eine Kette konstruieren.

````python
chain = prompt | llm | output_parser
chain.invoke({"text": product_description, "format_instructions": format_instructions})
````

## Anwendung zur Textextraktion

Sprachmodellbasiertes Lernen, häufig als LLM abgekürzt, hat zahlreiche Anwendungen in der Geschäftswelt. Eine häufige Anwendung von LLM ist die Textextraktion. Bei der Textextraktion geht es darum, bestimmte Informationen aus einem größeren Textkörper abzurufen. In Szenarien, in denen ein Datensatz beispielsweise verschiedene Informationen über Einzelpersonen enthält – von Geburtstagen bis hin zu Angaben zum Beruf – kann man LLM einsetzen, um sich nur auf die Geburtstage zu konzentrieren und irrelevante Daten effizient herauszufiltern. Die Stärke von LLM liegt in seiner Fähigkeit, Kontext zu erkennen und relevante Details basierend auf den Anforderungen des Benutzers zu extrahieren, wie der Code zeigt, der Geburtstagsdetails geschickt identifiziert und extrahiert, während er andere Einzelheiten außer Acht lässt.

````python
from langchain.prompts import ChatPromptTemplate

PROMPT = """
You are to extract any birthdays from the provided text, return the " \
date in the form 10-FEB-1990, or NONE if no birthday.

text: {text}"""

prompt_template = ChatPromptTemplate.from_template(PROMPT)

INPUT = "John was born on June 14, 1995, he was married on May 8, 2015."

chain = prompt_template | llm

result = chain.invoke({"text": INPUT})
print(result.content)
````

Derselbe Code kann eine Reihe von Textzeichenfolgen verarbeiten. Die Daten in diesen Zeichenfolgen liegen in verschiedenen Formaten vor. Das LLM kann die benötigten Geburtstage analysieren und finden und andere Informationen ignorieren. Beachten Sie, dass das Datum manchmal nicht wie gewünscht formatiert ist oder mehrere Daten zurückgegeben werden. Bald werden wir etwas über Prompt Engineering erfahren, das einige dieser Probleme löst.

````python
LIST = [
    "Anna started her first job on 15th January 2012. She was born on March 5, 1990.",
    "On 04/14/2007, Michael graduated from college. He was born on 20th July 1985.",
    "Born on 22nd October 1992, Sophia got married on 11.11.2016.",
    "Graduating from high school on June 5, 2005, was a big moment for Lucas. His birth date is 02/17/1987.",
    "Isabelle began her professional journey on 01/09/2016, having been born on December 3, 1994.",
    "Liam was born on May 12, 1988. He celebrated his wedding on 07-15-2014.",
    "Eva celebrated her college graduation on 20-05-2013. Her birthday falls on April 25, 1991.",
    "In 2006, specifically on 03.03.2006, Daniel started his first job. He came into this world on January 8, 1984.",
    "On 05.25.2011, Emily donned her graduation gown. Her birthdate is September 16, 1993.",
    "Henry marked his birthday on 11/30/1989. He tied the knot on October 10, 2017.",
]

for item in LIST:
    response = chain.invoke({"text": item})

    print(response.content)
````
