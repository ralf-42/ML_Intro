# Teil 2.2: Kategoriale und kontinuierliche Werte

Neuronale Netzwerke benötigen als Eingabe eine feste Anzahl von Spalten. Dieses Eingabeformat ist Tabellendaten sehr ähnlich; es muss vollständig numerisch sein. Es ist wichtig, die Daten so darzustellen, dass das neuronale Netzwerk damit trainieren kann. Bevor wir uns bestimmte Möglichkeiten zur Vorverarbeitung von Daten ansehen, ist es wichtig, vier grundlegende Datentypen zu berücksichtigen, wie sie durch [[Cite:stevens1946theory]](http://psychology.okstate.edu/faculty/jgrice/psyc3214/Stevens_FourScales_1946.pdf) definiert werden. Statistiker bezeichnen sie häufig als [[Cite:stevens1946theory]](http://psychology.okstate.edu/faculty/jgrice/psyc3214/Stevens_FourScales_1946.pdf):

- Zeichendaten (Zeichenfolgen)
- **Nominal** - Einzelne diskrete Elemente, keine Reihenfolge. Beispielsweise Farbe, Postleitzahl und Form.
- **Ordinalzahl** - Einzelne unterschiedliche Elemente haben eine implizite Reihenfolge. Zum Beispiel Gehaltsstufe, Berufsbezeichnung, Starbucks(tm)-Kaffeegröße (groß, vente, grande)
- Numerische Daten
- **Intervall** - Numerische Werte, kein definierter Start. Zum Beispiel Temperatur. Sie würden nie sagen: „Gestern war es doppelt so heiß wie heute.“
- **Verhältnis** - Numerische Werte, klar definierter Anfang. Zum Beispiel Geschwindigkeit. Man könnte sagen: „Das erste Auto fährt doppelt so schnell wie das zweite.“


## Kontinuierliche Werte kodieren

Eine häufige Transformation besteht darin, die Eingaben zu normalisieren. Manchmal ist es sinnvoll, numerische Eingaben in eine Standardform zu normalisieren, damit das Programm diese beiden Werte problemlos vergleichen kann. Stellen Sie sich vor, ein Freund sagt Ihnen, er habe einen Rabatt von 10 Dollar bekommen. Ist das ein gutes Geschäft? Vielleicht. Aber der Preis ist nicht normalisiert. Wenn Ihr Freund ein Auto gekauft hat, ist der Rabatt nicht so gut. Wenn Ihr Freund ein Mittagessen gekauft hat, ist das ein hervorragender Rabatt!

Prozentsätze sind eine weit verbreitete Form der Normalisierung. Wenn Ihr Freund Ihnen sagt, dass er 10 % Rabatt bekommen hat, wissen wir, dass das ein besserer Rabatt ist als 5 %. Dabei spielt es keine Rolle, wie hoch der Kaufpreis war. Eine weit verbreitete Normalisierungsmethode des maschinellen Lernens ist der Z-Score:

$$ z = \frac{x - \mu}{\sigma} $$

Um den Z-Score zu berechnen, müssen Sie auch den Mittelwert (&mu; oder $\bar{x}$) und die Standardabweichung (&sigma;) berechnen. Sie können den Mittelwert mit dieser Gleichung berechnen:

$$ \mu = \bar{x} = \frac{x_1+x_2+\cdots +x_n}{n} $$

Die Standardabweichung wird wie folgt berechnet:

$$ \sigma = \sqrt{\frac{1}{N} \sum\_{i=1}^N (x_i - \mu)^2} $$

Der folgende Python-Code ersetzt den MPG durch einen Z-Score. Autos mit durchschnittlichem MPG liegen nahe Null, über Null ist über dem Durchschnitt und unter Null ist unter dem Durchschnitt. Z-Scores von mehr als 3 über oder unter sind sehr selten; das sind Ausreißer.


```python
import pandas as pd
from scipy.stats import zscore


df = read_csv(
    "https://data.heatonresearch.com/data/t81-558/auto-mpg.csv", na_values=["NA", "?"]
)

set_option("display.max_columns", 7)
set_option("display.max_rows", 5)

df["mpg"] = zscore(df["mpg"])
display(df)
```

### Entfernen der ersten Ebene

Die Funktion **concat** enthält auch einen Parameter namens _drop_first_, der angibt, ob durch Entfernen der ersten Ebene k-1 Dummies aus k kategorialen Ebenen entfernt werden sollen. Warum sollten Sie die erste Ebene entfernen wollen, in diesem Fall _area_a_? Diese Technik bietet eine effizientere Kodierung, indem sie die normalerweise ungenutzte Kodierung von [0,0,0] verwendet. Wir kodieren den _area_ in nur drei Spalten und ordnen den kategorischen Wert von _a_ [0,0,0] zu. Der folgende Code demonstriert diese Technik.


```python
dummies = get_dummies(
    ["a", "b", "c", "d"], prefix="area", drop_first=True, dtype=int
)
print(dummies)
```

Wie Sie aus den obigen Daten ersehen können, fehlt die Spalte _area_a_, da **get_dummies** sie durch die Kodierung [0,0,0] ersetzt hat. Der folgende Code zeigt, wie diese Technik auf einen Datenrahmen angewendet wird.


```python
# Lesen des Datensatzes
df = read_csv(
    "https://data.heatonresearch.com/data/t81-558/jh-simple-dataset.csv",
    na_values=["NA", "?"],
)

# Kodieren Sie die Flächenspalte als Dummyvariablen
dummies = get_dummies(df["area"], drop_first=True, prefix="area")
df = concat([df, dummies], axis=1)
df.drop("area", axis=1, inplace=True)

# den codierten Datenrahmen anzeigen
set_option("display.max_columns", 0)
set_option("display.max_rows", 10)

display(df[["id", "job", "income", "area_b", "area_c", "area_d"]])
```

## Zielkodierung für kategorische Werte

Zielkodierung ist eine beliebte Technik für Kaggle-Wettbewerbe. Zielkodierung kann manchmal die Vorhersagekraft eines maschinellen Lernmodells erhöhen. Allerdings erhöht sie auch das Risiko einer Überanpassung dramatisch. Aufgrund dieses Risikos müssen Sie bei der Verwendung dieser Methode vorsichtig sein.

Im Allgemeinen kann die ZielKodierung nur dann auf ein kategoriales Merkmal angewendet werden, wenn die Ausgabe des maschinellen Lernmodells numerisch ist (Regression).

Das Konzept der Zielkodierung ist unkompliziert. Für jede Kategorie berechnen wir den durchschnittlichen Zielwert für diese Kategorie. Dann ersetzen wir zum Kodieren den Prozentsatz, der der Kategorie entspricht, die der kategorische Wert hat. Im Gegensatz zu Dummyvariablen, bei denen Sie für jede Kategorie mit Zielkodierung eine Spalte haben, benötigt das Programm nur eine einzige Spalte. Auf diese Weise ist die Zielkodierung effizienter als Dummyvariablen.


```python
import numpy as np

# Erstellen Sie einen kleinen Beispieldatensatz
np.random.seed(43)
df = DataFrame(
    {
        "cont_9": np.random.rand(10) * 100,
        "cat_0": ["dog"] * 5 + ["cat"] * 5,
        "cat_1": ["wolf"] * 9 + ["tiger"] * 1,
        "y": [1, 0, 1, 1, 1, 1, 0, 0, 0, 0],
    }
)

set_option("display.max_columns", 0)
set_option("display.max_rows", 0)
display(df)
```

Wir möchten sie in eine Zahl umwandeln, anstatt Dummy-Variablen für „Hund“ und „Katze“ zu erstellen. Wir möchten sie in eine Zahl umwandeln. Wir könnten 0 für eine Katze und 1 für einen Hund verwenden. Wir können jedoch mehr Informationen als nur das kodieren. Die einfache 0 oder 1 würde auch nur für ein Tier funktionieren. Überlegen Sie, was der mittlere Zielwert für Katze und Hund ist.


```python
means0 = df.groupby("cat_0")["y"].mean().to_dict()
means0
```

Die Gefahr besteht darin, dass wir jetzt den Zielwert ($y$) für das Training verwenden. Diese Technik führt möglicherweise zu Überanpassung. Die Möglichkeit einer Überanpassung ist noch größer, wenn eine kleine Anzahl einer bestimmten Kategorie vorhanden ist. Um dies zu verhindern, verwenden wir einen Gewichtungsfaktor. Je stärker das Gewicht, desto mehr Kategorien mit weniger Werten tendieren zum Gesamtdurchschnitt von $y$. Sie können diese Berechnung wie folgt durchführen.


```python
df["y"].mean()
```

Sie können die Zielkodierung wie folgt implementieren. Weitere Informationen zur Zielkodierung finden Sie im Artikel ["Target Encoding Done the Right Way"](https://maxhalford.github.io/blog/target-encoding/), auf dem dieser Code basiert.


```python
def calc_smooth_mean(df1, df2, cat_name, target, weight):
    # Berechnen Sie den globalen Mittelwert
    mean = df[target].mean()

    # Berechnen Sie die Anzahl der Werte und den Mittelwert jeder Gruppe
    agg = df.groupby(cat_name)[target].agg(["count", "mean"])
    counts = agg["count"]
    means = agg["mean"]

    # Berechnen Sie die „geglätteten“ Mittelwerte
    smooth = (counts * means + weight * mean) / (counts + weight)

    # Ersetzen Sie jeden Wert durch den entsprechenden geglätteten Mittelwert
    if df2 is None:
        return df1[cat_name].map(smooth)
    else:
        return df1[cat_name].map(smooth), df2[cat_name].map(smooth.to_dict())
```

Der folgende Code kodiert diese beiden Kategorien.


```python
WEIGHT = 5
df["cat_0_enc"] = calc_smooth_mean(
    df1=df, df2=None, cat_name="cat_0", target="y", weight=WEIGHT
)
df["cat_1_enc"] = calc_smooth_mean(
    df1=df, df2=None, cat_name="cat_1", target="y", weight=WEIGHT
)

set_option("display.max_columns", 0)
set_option("display.max_rows", 0)

display(df)
```

## Kodierung kategorialer Werte als Ordinalzahlen

Immer wenn die kategorischen Variablen geordnet sind, sollte eine Zahl verwendet werden. Die obige Liste hat 21 Ebenen und würde 21 Dummyvariablen zum Kodieren benötigen. Wenn man dies jedoch einfach in Dummyvariablen kodiert, geht die Reihenfolgeinformation verloren.
