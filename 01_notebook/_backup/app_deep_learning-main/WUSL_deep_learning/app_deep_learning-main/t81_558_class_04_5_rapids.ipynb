## Deep Learning und tabellarische Daten: Nicht immer eine perfekte Kombination

Obwohl Deep Learning in zahlreichen Bereichen, wie etwa der Bilderkennung, der Verarbeitung natürlicher Sprache und sogar beim Spielen von Brettspielen, tiefgreifende Veränderungen mit sich gebracht hat, greift es bei der Anwendung auf tabellarische Daten manchmal zu kurz, insbesondere im Vergleich zu herkömmlichen Algorithmen des maschinellen Lernens.

Es ist wichtig zu verstehen, dass neuronale Netzwerke, der Kern des Deep Learning, durchaus in der Lage sind, Regressions- und Klassifizierungsaufgaben für tabellarische Daten durchzuführen. Durch die Anwendung verschiedener Netzwerkarchitekturen und die Feinabstimmung von Hyperparametern können Sie ein Deep-Learning-Modell durchaus anhand solcher Datensätze trainieren.

In vielen Fällen übertreffen traditionellere Modelle wie XGBoost, Random Forests und andere Gradient Boosted Machines (GBMs) jedoch Deep-Learning-Modelle bei rein tabellarischen Datensätzen. Diese traditionellen Modelle haben ihre Effizienz bei der Verarbeitung tabellarischer Daten unter Beweis gestellt, da sie bestimmte Arten von Beziehungen und Abhängigkeiten zwischen Variablen besser erfassen können, die neuronalen Netzwerken manchmal entgehen.

Beweise für diesen Trend aus der Praxis finden sich in den Ergebnissen verschiedener Kaggle-Wettbewerbe. Kaggle, eine Plattform für Wettbewerbe im Bereich prädiktive Modellierung und Analytik, bietet eine Fülle von Daten zur Leistung von Modellen des maschinellen Lernens. Viele Gewinner dieser Wettbewerbe, insbesondere derjenigen, die sich auf tabellarische Daten konzentrieren, verwenden häufig GBMs. Beispielsweise war der XGBoost-Algorithmus, eine skalierbare und genaue Implementierung von Gradient-Boosting-Maschinen, Teil zahlreicher Gewinnerlösungen.

Ein Grund dafür ist, dass tabellarische Daten oft strukturierte und hierarchische Beziehungen aufweisen und Gradient-Boosting-Algorithmen diese Interaktionen hervorragend erfassen können. Andererseits könnten Deep-Learning-Modelle, insbesondere solche ohne spezifische Architekturentwürfe, mit solchen Daten Schwierigkeiten haben, wenn sie nicht mit großen Mengen an Trainingsdaten und sorgfältig ausgewählten Features ausgestattet werden.

Das bedeutet nicht, dass Deep Learning bei der Verarbeitung tabellarischer Daten vernachlässigt werden sollte. Vielmehr ist es wichtig, die Stärken und Schwächen der einzelnen Ansätze zu verstehen. In einigen Fällen kann die Kombination von Deep Learning mit herkömmlichen Methoden zu sehr leistungsfähigen Modellen führen. Für die meisten tabellarischen Datensätze haben sich die herkömmlichen Modelle jedoch als zuverlässiger, genauer und rechnerisch effizienter erwiesen.

Man darf nicht vergessen, dass es beim maschinellen Lernen keine Universallösung gibt. Die Wirksamkeit eines Modells hängt von der Art des Problems, der Qualität und Quantität der Daten sowie den verfügbaren Rechenressourcen ab. Als Praktiker besteht unsere Aufgabe darin, uns in diesem Umfeld zurechtzufinden und die besten Modelle für unsere spezifischen Anforderungen auszuwählen oder zu entwickeln.

## NVIDIA RAPIDS und Random Forests für tabellarische Vorhersagen

Da wir eine vollständig GPU-basierte Pipeline haben, werden wir ein vorheriges PyTorch-Beispiel für neuronale Netzwerke anpassen, um NVIDIA RAPIDS zu nutzen. Neuronale Netzwerke haben zwar ihre Stärken, es gibt jedoch Fälle, in denen traditionelle Modelle wie Random Forests sie übertreffen können, insbesondere bei der Verarbeitung bestimmter Arten tabellarischer Daten. In unserem Fall haben wir festgestellt, dass ein Random Forest-Modell für den Datensatz, mit dem wir arbeiten, die Leistung des zuvor verwendeten neuronalen Netzwerks tatsächlich übertrifft.

In diesem Abschnitt führen wir Sie Schritt für Schritt durch den Ersatz unseres neuronalen Netzwerks durch einen Random Forest und demonstrieren die Flexibilität und Vielseitigkeit, die RAPIDS bei der Auswahl und Änderung von Modellen für maschinelles Lernen bietet.

Ein Highlight dieser Anpassung wird die Verwendung der cudf-Bibliothek von RAPIDS sein. Dies ist eine leistungsstarke GPU-beschleunigte Datenmanipulationsbibliothek, mit der wir die gesamte erforderliche Vorverarbeitung unseres Datensatzes vollständig im GPU-Speicher durchführen können. Einer der wesentlichen Vorteile besteht darin, dass wir unsere verarbeiteten Daten direkt an das Random Forest-Modell übergeben können, ohne den GPU-Speicher zu verlassen, was die Effizienz unserer Data Science-Pipeline verbessert.

Dieser Code wird hier demonstriert.


```python
import cudf as pd
from cuml import train_test_split
from cuml.ensemble import RandomForestRegressor
from cuml.metrics import mean_squared_error
from cuml.preprocessing import StandardScaler


# Lesen des Datensatzes
df = read_csv(
    "https://data.heatonresearch.com/data/t81-558/jh-simple-dataset.csv",
    na_values=["NA", "?"],
)

# Dummies für den Job generieren
df = concat([df, get_dummies(df["job"], prefix="job", dtype=int)], axis=1)
df.drop("job", axis=1, inplace=True)

# Dummies für den Bereich generieren
df = concat([df, get_dummies(df["area"], prefix="area", dtype=int)], axis=1)
df.drop("area", axis=1, inplace=True)

# Dummies für Produkte generieren
df = concat([df, get_dummies(df["product"], prefix="product", dtype=int)], axis=1)
df.drop("product", axis=1, inplace=True)

# Fehlende Werte für Einkommen
med = df["income"].median()
df["income"] = df["income"].fillna(med)

# Bereiche standardisieren
scaler = StandardScaler()
df["income"] = scaler.fit_transform(df["income"].to_frame())
df["aspect"] = scaler.fit_transform(df["aspect"].to_frame())
df["save_rate"] = scaler.fit_transform(df["save_rate"].to_frame())
df["subscriptions"] = scaler.fit_transform(df["subscriptions"].to_frame())

# In cuDF DataFrame konvertieren
x_columns = df.columns.drop(["age", "id"])
x = df[x_columns]
y = df["age"]
x = x.astype("float32")
y = y.astype("float32")

# Legen Sie einen Zufallsstartwert für die Reproduzierbarkeit fest
import numpy as np

np.random.seed(42)

# Aufteilung in Trainings- und Testdatensätze
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)

# Trainieren Sie das Modell mit RandomForestRegressor
model = RandomForestRegressor(n_estimators=500, random_state=42, n_streams=1)
model.fit(x_train, y_train)

# Testen des Modells
y_pred = model.predict(x_test)

# RMSE berechnen
score = np.sqrt(mean_squared_error(y_test, y_pred))

print(f"Endergebnis (RMSE): {score}")
```
