# Teil 12.5: Zukünftige Richtungen des bestärkenden Lernens

Reinforcement Learning (RL) war zunächst sehr vielversprechend, da es Menschen in Spielen wie Schach und Go besiegte. Die 1990er Jahre läuteten die Geburt von RL ein, eine Zeit, in der das Feld ein Zusammenschluss einiger leidenschaftlicher Seelen war. 1995 begann sich beim ersten RL-Workshop der National Science Foundation die erste Community zu bilden.

Mit Anbruch des neuen Jahrtausends wuchs die Präsenz von RL still und leise, aber stetig, und schaffte es bis zum Aufkommen von DeepMind nie wirklich in die Mainstream-Maschinenlernforschung. Ihre innovative Synthese von Deep Learning mit RL, die sie mit Atari-Spielen demonstrierte, war eine Offenbarung. Sie löste eine Renaissance aus und machte RL plötzlich zum Objekt der Begierde für Technologiekonzerne und Startups, wie die fürstliche Übernahme von DeepMind durch Google beweist.

Für viele scheint RL ein Triumph zu sein, der seinen Siegeszug fortsetzen wird. Auf KI- und ML-Konferenzen gibt es jede Menge Sitzungen, die RL gewidmet sind, und die Flut an Beiträgen ist ungebrochen und untermauert die Erfolgsgeschichte.

Einige Forscher glauben jedoch, dass RL in einer Sackgasse steckt. Wenn wir RL die strengen Kriterien des Lernens in Echtzeit auferlegen, die in der realen Welt üblich sind, ändert sich das Bild. Die aktuellen Methoden des Deep RL, die in simulierten Realitäten glänzen, in denen Fehler keine Rolle spielen und Wiederholungen unendlich sind, scheitern an dem komplexen Lernen, das das reale Leben erfordert. Denken Sie daran, wie Menschen lernen, die Kunst des Autofahrens zu meistern. Es ist nicht nur eine Funktion des Bedienens des Fahrzeugs, sondern die Kulmination jahrelanger passiver Beobachtung, ein komplexer Tanz der Sinne und der Wahrnehmung, den kein aktueller RL-Algorithmus nachbilden kann. OpenAI, einst der führende Befürworter von RL, hat sich vom OpenAI Gym getrennt.

Bei den Versuchen, Maschinen durch RL zu lehren, wurde dieser grundlegende Aspekt des menschlichen Lernens oft außer Acht gelassen – der riesige Vorrat an implizitem Wissen, den wir in jede neue Lernerfahrung einbringen. Der starke Kontrast in der Lerneffizienz zwischen Menschen und RL-Algorithmen wird deutlich. Während Deep RL Tage und Millionen von Iterationen braucht, um ein Spiel wie Frostbite zu verstehen, braucht ein Mensch nur eine Minute und ein paar hundert Versuche. Dieser Unterschied darf nicht unterschätzt werden.

Wo bleibt also das RL? Kann es aus der potenziellen Sackgasse gerettet werden, in die die aktuellen Trends führen? Manche glauben, die Antwort liege nicht in der Aufgabe des RL, sondern in seiner Weiterentwicklung. Die Zukunft des RL sollte ein Modell umfassen, das Beobachtung, Nachahmung, Transferlernen und die Unterstützung durch Vorwissen integriert. Dieses Paradigma steht im Einklang mit der Art und Weise, wie Menschen komplexe Fähigkeiten erwerben. Die Überlegungen von Richard Feynman zur Bildung stimmen tief mit diesem Thema überein. Trotz seiner beispiellosen Eloquenz im Physikunterricht erkannte Feynman die Grenzen des Unterrichts. Diese Ansicht bezieht sich auf ein älteres, vielleicht universelleres Verständnis des Lernens: Wir nehmen am besten das auf, worauf wir bereits vorbereitet sind.

Diese Erkenntnis bringt uns zu einem entscheidenden Wendepunkt für RL. Der Weg in die Zukunft besteht nicht darin, den menschlichen Lernprozess nachzuahmen, sondern einen symbiotischen Rahmen zu schaffen, in dem Maschinen von der Fülle menschlicher Erfahrungen profitieren können. Wir stehen am Rande einer transformativen Entwicklung von RL oder seines Niedergangs und seiner Obsoleszenz. Die Wahl liegt bei uns, die Richtung liegt bei uns. Die Zukunft von RL wird davon abhängen, ob wir uns dafür entscheiden, aus der Tiefe menschlicher Erfahrungen zu lernen oder den Weg isolierter, wissensfreier Berechnungen weiter zu beschreiten.
