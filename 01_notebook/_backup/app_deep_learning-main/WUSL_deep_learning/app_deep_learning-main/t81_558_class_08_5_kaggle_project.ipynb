Das folgende Programm generiert eine Übermittlungsdatei für „Iris Kaggle“. Sie können sie als Ausgangspunkt für Aufgabe 3 verwenden.

```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset
from sklearn import metrics
import numpy as np

# Lesen Sie die Daten
df_train = read_csv(
    "https://data.heatonresearch.com/data/t81-558/datasets/kaggle_iris_train.csv",
    na_values=["NA", "?"],
)

# Merkmalsvektor kodieren
df_train.drop("id", axis=1, inplace=True)

num_classes = len(df_train.groupby("species").species.nunique())
print("Number of classes: {}".format(num_classes))

# In Numpy konvertieren - Klassifizierung
x = df_train[["sepal_l", "sepal_w", "petal_l", "petal_w"]].values
dummies = get_dummies(df_train["species"])  # Einstufung
species = dummies.columns
y = dummies.values

# Aufteilung in Trainieren/Testen
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.25, random_state=45
)

# In PyTorch-Tensoren konvertieren
x_train, y_train = torch.tensor(x_train, dtype=torch.float32), torch.tensor(
    y_train, dtype=torch.float32
)
x_test, y_test = torch.tensor(x_test, dtype=torch.float32), torch.tensor(
    y_test, dtype=torch.float32
)

# Definieren Sie das Modell mit torch.nn.Sequential
model = nn.Sequential(
    nn.Linear(x.shape[1], 50),
    nn.ReLU(),
    nn.Linear(50, 25),
    nn.Linear(25, y.shape[1]),
    nn.Softmax(dim=1),
)

optimizer = torch.optim.Adam(model.parameters())
loss_fn = nn.CrossEntropyLoss()

# Trainingsrunde mit frühzeitigem Stoppen
n_epochs = 1000
patience = 5
best_loss = float("inf")
early_stopping_counter = 0

for epoch in range(n_epochs):
    # Zug
    model.train()
    optimizer.zero_grad()
    y_pred = model(x_train)
    loss = loss_fn(y_pred, torch.argmax(y_train, 1))
    loss.backward()
    optimizer.step()

    # Bestätigen
    model.eval()
    with torch.no_grad():
        y_val_pred = model(x_test)
        val_loss = loss_fn(y_val_pred, torch.argmax(y_test, 1))

    if val_loss < best_loss:
        best_loss = val_loss
        early_stopping_counter = 0
    else:
        early_stopping_counter += 1
        if early_stopping_counter >= patience:
            print("Early Stopping!")
            break
```

Nachdem wir das neuronale Netzwerk trainiert haben, können wir seinen Log-Verlust überprüfen.

```python
# Berechnen Sie den Multi-Log-Loss-Fehler
model.eval()
with torch.no_grad():
    y_pred = model(x_test)
    y_pred = y_pred.numpy()
score = metrics.log_loss(y_test, y_pred)
print("Log loss score: {}".format(score))
```

* [kaggle_mpg_test.csv](https://data.heatonresearch.com/data/t81-558/datasets/kaggle_auto_test.csv) – Die Daten, anhand derer Kaggle Sie bewertet. Enthält nur Eingaben, Sie müssen Antworten angeben. (enthält x)
* [kaggle_mpg_train.csv](https://data.heatonresearch.com/data/t81-558/datasets/kaggle_auto_test.csv) – Die Daten, die Sie zum Trainieren verwenden werden. (enthält x und y)
* [kaggle_mpg_sample.csv](https://data.heatonresearch.com/data/t81-558/datasets/kaggle_auto_sample.csv) – Eine Beispielübermittlung für Kaggle. (enthält x und y)

Das folgende Programm generiert eine Übermittlungsdatei für „MPG Kaggle“.

```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
import pandas as pd
import io
import os
import requests
import numpy as np
from sklearn import metrics

# Herunterladen und Vorverarbeiten von Daten
save_path = "."
df = read_csv(
    "https://data.heatonresearch.com/data/t81-558/datasets/kaggle_auto_train.csv",
    na_values=["NA", "?"],
)
cars = df["name"]
df["horsepower"] = df["horsepower"].fillna(df["horsepower"].median())

x = df[
    [
        "cylinders",
        "displacement",
        "horsepower",
        "weight",
        "acceleration",
        "year",
        "origin",
    ]
].values
y = df["mpg"].values

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.25, random_state=42
)

# Konvertieren Sie Numpy-Arrays in PyTorch-Tensoren
x_train, x_test, y_train, y_test = map(torch.tensor, (x_train, x_test, y_train, y_test))
x_train, x_test = x_train.float(), x_test.float()
y_train, y_test = y_train.float().unsqueeze(1), y_test.float().unsqueeze(1)

# Definieren Sie das neuronale Netzwerk mit Sequential
model = nn.Sequential(
    nn.Linear(x_train.shape[1], 25),
    nn.ReLU(),
    nn.Linear(25, 10),
    nn.ReLU(),
    nn.Linear(10, 1),
)

# Definieren Sie Verlust und Optimierer
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters())

# Kriterien für die vorzeitige Beendigung
min_delta = 1e-3
patience = 5
best_loss = float("inf")
count = 0

# Trainingsschleife
for epoch in range(1000):
    model.train()
    optimizer.zero_grad()
    outputs = model(x_train)
    loss = criterion(outputs, y_train)
    loss.backward()
    optimizer.step()

    with torch.no_grad():
        model.eval()
        val_outputs = model(x_test)
        val_loss = criterion(val_outputs, y_test)
        if val_loss < best_loss - min_delta:
            best_loss = val_loss
            count = 0
        else:
            count += 1
        if count > patience:
            print("Early stopping")
            break
```

Nachdem wir das neuronale Netzwerk trainiert haben, können wir seinen RMSE-Fehler überprüfen.

```python
# Vorhersagen
model.eval()
with torch.no_grad():
    pred = model(x_test)

# RMSE messen
score = torch.sqrt(criterion(pred, y_test))
print("Final score (RMSE):", score.item())
```
