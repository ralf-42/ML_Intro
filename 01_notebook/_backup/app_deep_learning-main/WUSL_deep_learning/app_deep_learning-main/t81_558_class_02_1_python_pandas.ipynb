## Fehlende Werte

Fehlende Werte sind eine Realität des maschinellen Lernens. Im Idealfall enthält jede Datenzeile Werte für alle Spalten. Dies ist jedoch selten der Fall. Die meisten Werte sind in der MPG-Datenbank vorhanden. In der PS-Spalte fehlen jedoch Werte. Eine gängige Vorgehensweise besteht darin, fehlende Werte durch den Medianwert für diese Spalte zu ersetzen. Das Programm berechnet [median](https://en.wikipedia.org/wiki/Median). Der folgende Code ersetzt alle NA-Werte in PS durch den Medianwert:


```python
df = read_csv(
    "https://data.heatonresearch.com/data/t81-558/auto-mpg.csv", na_values=["NA", "?"]
)
print(f"PS hat n/a? {isnull(df['horsepower']).values.any()}")

print("Filling missing values...")
med = df["horsepower"].median()
df["horsepower"] = df["horsepower"].fillna(med)
# df = df.dropna() # Sie können NA-Werte auch einfach löschen

print(f"PS hat n/a? {isnull(df['horsepower']).values.any()}")
```

## Umgang mit Ausreißern

Ausreißer sind Werte, die ungewöhnlich hoch oder niedrig sind. Wir betrachten Ausreißer normalerweise als Werte, die mehrere Standardabweichungen vom Mittelwert entfernt sind. Manchmal sind Ausreißer einfach Fehler; dies ist ein Ergebnis von [observation error](https://en.wikipedia.org/wiki/Observational_error). Ausreißer können auch wirklich große oder kleine Werte sein, die möglicherweise schwer zu beheben sind. Die folgende Funktion kann solche Werte entfernen.


```python
# Entfernen Sie alle Zeilen, bei denen die angegebene Spalte +/- SD-Standardabweichungen beträgt
def remove_outliers(df, name, sd):
    drop_rows = df.index[(np.abs(df[name] - df[name].mean()) >= (sd * df[name].std()))]
    df.drop(drop_rows, axis=0, inplace=True)
```

Der folgende Code löscht alle Zeilen aus dem Auto-MPG-Datensatz, bei denen die Leistung zwei oder mehr Standardabweichungen über oder unter dem Mittelwert liegt.


```python
import numpy as np


df = read_csv(
    "https://data.heatonresearch.com/data/t81-558/auto-mpg.csv", na_values=["NA", "?"]
)

# Merkmalsvektor erstellen
med = df["horsepower"].median()
df["horsepower"] = df["horsepower"].fillna(med)

# Löschen der Namensspalte
df.drop("name", 1, inplace=True)

# Ausreißer in der Leistung eliminieren
print("Length before MPG outliers dropped: {}".format(len(df)))
remove_outliers(df, "mpg", 2)
print("Length after MPG outliers dropped: {}".format(len(df)))

set_option("display.max_columns", 0)
set_option("display.max_rows", 5)
display(df)
```

## Training und Validierung

Wir müssen ein maschinelles Lernmodell anhand seiner Fähigkeit bewerten, Werte vorherzusagen, die es noch nie zuvor gesehen hat. Aus diesem Grund teilen wir die Trainingsdaten häufig in einen Validierungs- und einen Trainingssatz auf. Das maschinelle Lernmodell lernt aus den Trainingsdaten, wird aber letztendlich auf der Grundlage der Validierungsdaten bewertet.

- **Trainingsdaten** - **In Beispieldaten** – Die Daten, die das neuronale Netzwerk zum Trainieren verwendet hat.
- **Validierungsdaten** – **Out-of-Sample-Daten** – Die Daten, anhand derer das maschinelle Lernmodell ausgewertet wird, nachdem es an die Trainingsdaten angepasst wurde.

Es gibt zwei effektive Möglichkeiten, mit Trainings- und Validierungsdaten umzugehen:

- **Training/Validierung-Aufteilung** - Das Programm teilt die Daten nach einem bestimmten Verhältnis zwischen einem Trainings- und einem Validierungssatz (Hold-out) auf. Typische Raten sind 80 % Training und 20 % Validierung.
- **K-fach-Kreuzvalidierung** - Das Programm teilt die Daten in mehrere Faltungen und Modelle auf. Da das Programm die gleiche Anzahl an Modellen wie Faltungen erstellt, kann das Programm Out-of-Sample-Vorhersagen für den gesamten Datensatz generieren.

Der folgende Code teilt die MPG-Daten in einen Trainings- und einen Validierungssatz auf. Der Trainingssatz verwendet 80 % der Daten und der Validierungssatz 20 %. Abbildung 2.TRN-VAL zeigt, wie wir ein Modell mit 80 % der Daten trainieren und es dann anhand der verbleibenden 20 % validieren.

**Abbildung 2.TRN-VAL: Training und Validierung**
![Training and Validation](https://raw.githubusercontent.com/jeffheaton/t81_558_deep_learning/master/images/class_1_train_val.png "Training and Validation")

## Speichern eines Dataframes in Pickle

Eine Vielzahl von Softwareprogrammen kann als CSV gespeicherte Textdateien verwenden. Allerdings dauert ihre Generierung länger und bei der Konvertierung kann manchmal etwas an Genauigkeit verloren gehen. Im Allgemeinen werden Sie im CSV-Format ausgeben, da es sehr kompatibel ist, auch außerhalb von Python. Ein anderes Format ist [Pickle](https://docs.python.org/3/library/pickle.html). Der folgende Code speichert den Dataframe in Pickle. Pickle speichert Daten in der exakten Binärdarstellung, die von Python verwendet wird. Der Vorteil ist, dass beim Wechsel ins CSV-Format keine Daten verloren gehen. Der Nachteil ist, dass im Allgemeinen nur Python-Programme Pickle-Dateien lesen können.


```python
import os
import pickle

path = "."

df = read_csv(
    "https://data.heatonresearch.com/data/t81-558/auto-mpg.csv", na_values=["NA", "?"]
)

filename_write = os.path.join(path, "auto-mpg-shuffle.pkl")
df = df.reindex(np.random.permutation(df.index))

with open(filename_write, "wb") as fp:
    pickle.dump(df, fp)
```
