## Was ist Deep Learning

Der Schwerpunkt dieses Buches liegt auf Deep Learning, einer weit verbreiteten Art des maschinellen Lernens, die auf den ursprünglichen neuronalen Netzwerken aufbaut, die in den 1980er Jahren populär wurden. Es gibt kaum einen Unterschied zwischen der Berechnung eines tiefen neuronalen Netzwerks und dem ersten neuronalen Netzwerk. Wir konnten schon immer tiefe neuronale Netzwerke erstellen und berechnen. Ein tiefes neuronales Netzwerk ist nichts anderes als ein neuronales Netzwerk mit vielen Schichten. Obwohl wir schon immer tiefe neuronale Netzwerke erstellen/berechnen konnten, fehlte uns ein wirksames Mittel, um sie zu trainieren. Deep Learning bietet ein effizientes Mittel, um tiefe neuronale Netzwerke zu trainieren.


Wenn Deep Learning eine Art maschinelles Lernen ist, stellt sich die Frage: „Was ist maschinelles Lernen?“ Abbildung 1.ML-DEV zeigt, wie sich maschinelles Lernen von der traditionellen Softwareentwicklung unterscheidet.

## Regression, Klassifizierung und mehr

Die Forschung zum maschinellen Lernen befasst sich mit Problemen im weitesten Sinne des überwachten und unüberwachten Lernens. Überwachtes Lernen findet statt, wenn Sie das richtige Ergebnis für jedes Element im Trainingsset kennen. Unüberwachtes Lernen hingegen verwendet Trainingssets, bei denen kein richtiges Ergebnis bekannt ist. Deep Learning unterstützt sowohl überwachtes als auch unüberwachtes Lernen, fügt jedoch auch Verstärkung und kontroverses Lernen hinzu. Verstärkungslernen bringt dem neuronalen Netzwerk bei, Aktionen basierend auf einer Umgebung auszuführen. Kontroverses Lernen lässt zwei neuronale Netzwerke gegeneinander antreten, um zu lernen, wann die Daten keine richtigen Ergebnisse liefern. Forscher fügen ständig neue Deep-Learning-Trainingstechniken hinzu.

Anwender des maschinellen Lernens unterteilen überwachtes Lernen normalerweise in Klassifizierung und Regression. Klassifizierungsnetzwerke können Finanzdaten akzeptieren und das Anlagerisiko als riskant oder sicher klassifizieren. In ähnlicher Weise gibt ein neuronales Regressionsnetzwerk eine Zahl aus und kann dieselben Daten verwenden und einen Risikowert zurückgeben. Darüber hinaus können neuronale Netzwerke mehrere Regressions- und Klassifizierungswerte gleichzeitig ausgeben.

Einer der leistungsfähigsten Aspekte neuronaler Netzwerke besteht darin, dass die Eingabe und Ausgabe eines neuronalen Netzwerks viele verschiedene Typen haben kann, wie zum Beispiel:

* Ein Bild
* Eine Zahlenreihe, die Text, Audio oder eine andere Zeitreihe darstellen könnte
* Eine Regressionszahl
* Eine Klassifizierungsklasse

## Warum Deep Learning?

Bei tabellarischen Daten weisen neuronale Netzwerke häufig keine wesentlich bessere Leistung auf als andere Modelle, beispielsweise:

* [Support Vector Machines](https://en.wikipedia.org/wiki/Support-vector_machine)
* [Random Forests](https://en.wikipedia.org/wiki/Random_forest)
* [Gradient Boosted Machines](https://en.wikipedia.org/wiki/Gradient_boosting)

Wie diese anderen Modelle können neuronale Netzwerke sowohl **Klassifizierung** als auch **Regression** durchführen. Bei der Anwendung auf relativ niedrigdimensionale tabellarische Datenaufgaben bieten tiefe neuronale Netzwerke nicht unbedingt eine signifikant höhere Genauigkeit als andere Modelltypen. Die meisten modernen Lösungen basieren jedoch bei Bildern, Videos, Texten und Audiodaten auf tiefen neuronalen Netzwerken.
