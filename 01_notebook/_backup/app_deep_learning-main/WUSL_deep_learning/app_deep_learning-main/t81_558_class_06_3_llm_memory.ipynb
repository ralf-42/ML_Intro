# Teil 6.3: LLM-Speicher

Das menschliche Gehirn verfügt sowohl über ein Langzeitgedächtnis als auch ein Kurzzeitgedächtnis. Das Langzeitgedächtnis ist das, was der Mensch im Laufe seines Lebens gelernt hat. Das Kurzzeitgedächtnis ist das, was der Mensch erst in der letzten Minute oder so entdeckt hat. Für den Menschen bedeutet Lernen, das Kurzzeitgedächtnis in ein Langzeitgedächtnis umzuwandeln, das wir behalten werden.

Dieser Prozess funktioniert bei einem LLM etwas anders. Das Langzeitgedächtnis war das Gewicht des neuronalen Netzwerks, als es anfänglich trainiert oder feinabgestimmt wurde. Das Kurzzeitgedächtnis sind zusätzliche Informationen, die das LLM von vorherigen Eingabeaufforderungen behalten soll. Wenn die erste Eingabeaufforderung beispielsweise „Mein Name ist Jeff“ lautet, wird das LLM Sie wahrscheinlich begrüßen und Ihren Namen wiederholen. Wenn die zweite Eingabeaufforderung jedoch „Wie heiße ich?“ lautet, wird das LLM die Antwort nicht kennen, ohne eine Speicherkomponente hinzuzufügen.

Diese Speicherobjekte, die LangChain bereitstellt, stellen eine Art Kurzzeitgedächtnis dar. Es ist wichtig zu beachten, dass diese Objekte das Langzeitgedächtnis des LLM nicht beeinflussen, und sobald Sie das Speicherobjekt löschen, vergisst das LLM es. Außerdem kann das Speicherobjekt nur eine begrenzte Menge an Informationen speichern; neuere Informationen können ältere Informationen ersetzen, sobald es gefüllt ist.

Ein wichtiger Punkt, den Sie bedenken sollten, ist, dass LLMs nur ihre Eingabeaufforderung haben. Um einen solchen Speicher bereitzustellen, hängen diese Objekte alles, was sich das LLM merken soll, an die Eingabeaufforderung an. In diesem Abschnitt werden zwei Möglichkeiten beschrieben, die Eingabeaufforderung mit vorherigen Informationen zu erweitern: ein Puffer und eine Zusammenfassung. Dem Puffer wird ein Skript der letzten Konversation bis zu diesem Punkt vorangestellt. Der Zusammenfassungsansatz hält einen ständig aktualisierten Zusammenfassungsabsatz der Konversation bereit.

## Konversationspuffer-Fensterspeicher

Die LangChain-Bibliothek enthält ein Konversationsobjekt namens **ConversationChain**; dieses Objekt ermöglicht eine fortlaufende Konversation mit einem LLM. Für jedes Konversationsobjekt müssen Sie auch einen Speicher angeben. Für dieses erste Beispiel verwenden wir das Objekt **ConversationBufferWindowMemory**. Dieses Objekt speichert eine Abschrift der letzten Konversation als Referenz. Dieser Speicher ermöglicht es dem Konversationsobjekt, sich zu merken, was Sie gefragt oder gesagt haben und welche Antworten es Ihnen gegeben hat.

````python
from langchain.chains import ConversationChain
from langchain.memory import ConversationBufferWindowMemory

memory = ConversationBufferWindowMemory()
conversation = ConversationChain(llm=llm, memory=memory, verbose=False)
````

Wir können jetzt ein Gespräch mit dem LLM führen.

````python
conversation.predict(input="Hi, my name is Jeff")
````

````python
conversation.predict(input="What is my name?")
````

Wir können uns anschauen, was der Speicher jetzt enthält.

````python
conversation.memory.load_memory_variables({})
````

## Benutzerdefinierte Konversationsbots

Sie können die Eingabeaufforderungsvorlage für einen Konversationsbot definieren. Mit dieser Technik können Sie einen Bot mit einem Namen erstellen und eine spezielle Aufgabe ausführen. In diesem Fall haben wir einen Bot mit dem Namen „WashU Assistant“ erstellt, den wir zur Unterstützung von Studenten konzipiert haben.

````python
# Jetzt können wir es überschreiben und auf „KI-Assistent“ setzen.
from langchain.prompts.prompt import PromptTemplate

template = """The following is a friendly conversation between a human and an
AI to assist Washington University Students. The AI should stick to topics
about Washington University. If the AI does not know the answer to a question,
it should suggest the student speak to their advisor.

Current conversation:
{history}
Human: {input}
WashU Assistant:"""
PROMPT = PromptTemplate(input_variables=["history", "input"], template=template)
conversation = ConversationChain(
    prompt=PROMPT,
    llm=llm,
    verbose=False,
    memory=ConversationBufferWindowMemory(ai_prefix="WashU Assistant"),
)
````

Wir können jetzt ein Gespräch mit dem WUSTL-Assistent-Bot führen.

````python
conversation.predict(input="Where is the bookstore?")
````

````python
conversation.predict(input="What is a nice quiet area to study?")
````

````python
conversation.predict(input="Which of these is closest to the bookstore?")
````

````python
conversation.predict(input="What is the meaning of life.")
````

Wir können uns anschauen, was der Speicher jetzt enthält.

````python
conversation.memory.load_memory_variables({})
````

## Gesprächszusammenfassungsspeicher

Sehen wir uns nun die Verwendung eines etwas komplexeren Speichertyps an, des ConversationSummaryMemory-Objekts. Dieser Speichertyp erstellt eine Zusammenfassung der Konversation im Zeitverlauf. Dieser Speicher kann hilfreich sein, um Informationen aus der Konversation im Zeitverlauf zu verdichten. Der Konversationszusammenfassungsspeicher fasst die Konversation zusammen und speichert die aktuelle Zusammenfassung im Speicher. Sie können diesen Speicher verwenden, um die bisherige Konversationszusammenfassung in eine Eingabeaufforderung/Kette einzufügen. Dieser Speicher ist besonders nützlich für längere Konversationen, bei denen das wörtliche Speichern des bisherigen Nachrichtenverlaufs in der Eingabeaufforderung zu viele Token beanspruchen würde.

````python
from langchain.memory import ConversationSummaryMemory

memory = ConversationSummaryMemory(llm=llm)
conversation = ConversationChain(llm=llm, memory=memory, verbose=False)
````

````python
conversation.predict(input="I am a computer scientist, what do you do for a living?")
````

Wir können uns anschauen, was der Speicher jetzt enthält.

````python
conversation.memory.load_memory_variables({})
````
