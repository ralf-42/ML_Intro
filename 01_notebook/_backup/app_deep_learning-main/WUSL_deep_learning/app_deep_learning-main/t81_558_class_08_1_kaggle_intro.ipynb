## Wie Kaggle-Wettbewerbe bewertet werden

Kaggle erhält vom Wettbewerbssponsor einen Datensatz, wie in Abbildung 8.SCORE zu sehen ist. Kaggle unterteilt diesen Datensatz wie folgt:

* **Vollständiger Datensatz** – Dies ist der vollständige Datensatz.
* **Trainingsdatensatz** – Dieser Datensatz stellt sowohl die Eingaben als auch die Ergebnisse für den Trainingsteil des Datensatzes bereit.
* **Testdatensatz** - Dieser Datensatz stellt die vollständigen Testdaten bereit, gibt jedoch nicht die Ergebnisse an. Ihre Übermittlungsdatei sollte die prognostizierten Ergebnisse für diesen Datensatz enthalten.
* **Öffentliche Bestenliste** - Kaggle sagt Ihnen nicht, welcher Teil des Testdatensatzes zur öffentlichen Bestenliste beiträgt. Ihr öffentlicher Score wird auf Grundlage dieses Teils des Datensatzes berechnet.
* **Private Bestenliste** - Ebenso sagt Ihnen Kaggle nicht, welcher Teil des Testdatensatzes zur öffentlichen Bestenliste beiträgt. Ihr endgültiger Punktestand/Rang wird auf Grundlage dieses Teils berechnet. Sie sehen Ihren privaten Bestenlistenpunktstand erst am Ende.

## Vorbereiten einer Kaggle-Einreichung

Sie übermitteln den Code Ihrer Lösung nicht an Kaggle. Bei Wettbewerben werden Sie ausschließlich auf der Grundlage der Genauigkeit Ihrer Übermittlungsdatei bewertet. Eine Kaggle-Übermittlungsdatei ist immer eine CSV-Datei, die die **ID** der Zeile enthält, die Sie vorhersagen, und die Antwort. Für den Titanic-Wettbewerb sieht eine Übermittlungsdatei ungefähr so ​​aus:

```
PassengerId,Survived
892,0
893,1
894,1
895,0
896,0
897,1
...
```

Die obige Datei enthält die Vorhersage für jeden einzelnen Passagier. Sie sollten nur Vorhersagen für IDs machen, die in der Testdatei enthalten sind. Ebenso sollten Sie für jede Zeile in der Testdatei eine Vorhersage machen. Bei manchen Wettbewerben sind die Antworten in unterschiedlichem Format. Eine Mehrfachklassifizierung hat beispielsweise normalerweise eine Spalte für jede Klasse und Ihre Vorhersagen für jede Klasse.
