<p><font color='black' size="5">📈 Algorithmus: LSTM – Zeitreihenvorhersage</font></p>

LSTM-Netze (Long Short-Term Memory) sind neuronale Netze, die für sequenzielle Daten – wie Zeitreihen – besonders geeignet sind. Eine LSTM-Zelle kann relevante Informationen aus der Vergangenheit behalten und unwichtige vergessen.

**Prinzip:** Ein Zeitfenster vergangener Werte wird als Eingabe verwendet, um den nächsten Wert vorherzusagen.

**Wichtige Konzepte:**

| Begriff | Bedeutung |
|---------|----------|
| Zeitfenster | Anzahl der vergangenen Schritte als Eingabe |
| LSTM-Zelle | Speichert und verarbeitet sequenzielle Informationen |
| Vorwärtsprognose | Vorhersage eines oder mehrerer zukünftiger Werte |

**In der Praxis relevant wenn:**
- Zeitreihen vorhergesagt werden sollen und die Reihenfolge sowie der historische Kontext für die Prognose entscheidend sind
- Langfristige Abhängigkeiten in Sequenzdaten modelliert werden sollen — LSTM kann relevante Informationen aus der Vergangenheit behalten und unwichtige vergessen
- Klassische statistische Modelle (ARIMA) nicht ausreichen, weil nichtlineare Muster oder mehrere saisonale Zyklen vorliegen

**Nicht geeignet wenn:**
- Sehr kurze Zeitreihen (< 50 Datenpunkte) vorliegen → LSTMs brauchen ausreichend historischen Kontext für das Training
- Einfache, stationäre Zeitreihen vorliegen → dann ARIMA oder Exponential Smoothing sind effizienter

**Typischer Fehler:**
`shuffle=True` beim Train-Test-Split für Zeitreihendaten verwenden — das zerstört die zeitliche Reihenfolge und der LSTM lernt aus zukünftigen Daten. Zeitreihen immer chronologisch aufteilen: `shuffle=False`.

```python
#@markdown   <p><font size="4" color='green'>  LSTM – Zeitreihenvorhersage</font> </br></p>

import base64
from IPython.display import Image, display

diagram = """
flowchart LR
    TS[Zeitreihe t-n bis t] --> WIN[Zeitfenster bilden]
    WIN --> LSTM[LSTM-Schicht]
    LSTM --> DENSE[Dense-Schicht]
    DENSE --> PRED[Vorhersage t+1]
"""

encoded = base64.urlsafe_b64encode(diagram.strip().encode()).decode()
display(Image(url=f"https://mermaid.ink/img/{encoded}", width=950))
```

<p><font color='black' size="5">
Zufallszahl initialisieren
</font></p>

```python
set_random_seed(42)
enable_op_determinism()
```

<p><font color='black' size="5">
Modellaufbau
</font></p>

[Core layers](https://keras.io/api/layers/core_layers/)    
[Recurrent layers](https://keras.io/api/layers/recurrent_layers/)   
[Layer activation functions](https://keras.io/api/layers/activations/)   


```python
model = Sequential(name="Zeitreihenanalyse")
model.add(Input(shape=(past_steps, num_features)))
model.add(LSTM(256, activation="relu", return_sequences=True))
model.add(LSTM(128, activation="relu", return_sequences=True))
model.add(LSTM(64, activation="relu", return_sequences=True))
model.add(LSTM(32, activation="relu", return_sequences=False))
model.add(Dense(1))
```

```python
model.summary()
```

```python
# Visualisierung neuronales Netz
plot_model(
    model,
    to_file="nn_structure.png",
    show_shapes=True,
    show_dtype=True,
    show_layer_names=True,
    dpi=100,
    expand_nested=True,
    show_layer_activations=True
)
```

```python
# Anzahl Parameter je Layer
for layer in model.layers:
    print(f"{layer.name}: {layer.count_params()} Parameter")
```


<p><font color='black' size="5">
Compile
</font></p>

```python
model.compile(optimizer="adam", loss="mse", metrics=["mae"])
```

<p><font color='black' size="5">
Callbacks
</font></p>

```python
early = EarlyStopping(monitor="mae", patience=2)

reduce_lr = ReduceLROnPlateau(
    monitor="loss",
    factor=0.1,
    patience=2,
    verbose=1,
    mode="auto",
    min_delta=0.0001,
    cooldown=0,
    min_lr=0,
)

check = ModelCheckpoint(filepath="model.keras", monitor="mae", save_best_only=True)
```

<p><font color='black' size="5">
Training
</font></p>

```python
model.fit(generator, epochs=20, batch_size=256, callbacks=[early, check])
```

```python
save_history = model.history.history
```

<p><font color='black' size="5">
Rollierende Prognose
</font></p>

```python
### Bestes Modell laden
model = load_model("model.keras")
```

```python
# Rollierende Vorhersage - Logik für eine "echte" Prognose in die Zukunft
initial_sequence_generator = TimeseriesGenerator(
    target_train.tail(past_steps + 1).values,
    target_train.tail(past_steps + 1).values,
    length=past_steps,
    batch_size=future_steps,
)

# x, y = initial_sequence_generator[0]
# print(x.shape, y.shape)

# Leere Liste für die Vorhersagen
target_test_pred = []
# Initialisierung für 1. Prognose
initial_sequence, y_dummy = initial_sequence_generator[0]

# Anzahl der Schritte, die Sie in die Zukunft vorhersagen möchten

for _ in range(target_test.shape[0]):
    # Machen Sie eine Vorhersage mit der aktuellen Sequenz
    current_prediction = model.predict(initial_sequence)

    # Fügen Sie die Vorhersage zu Ihrer Vorhersageliste hinzu
    target_test_pred.append(current_prediction[0, 0])

    # Aktualisieren Sie die Sequenz: Entfernen Sie den ältesten Punkt und fügen Sie die neue Vorhersage hinzu
    initial_sequence = np.roll(initial_sequence, -1, axis=1)
    initial_sequence[0, -1] = current_prediction
```

```python
cube = data[-len(target_test):]
cube["Real"] = target_test.values
cube["Pred"] = target_test_pred
```

<p><font color='black' size="5">
Bestimmtheitsmass
</font></p>

```python
f"Bestimmtheitsmass: {r2_score(cube['Real'].values, cube['Pred'].values):.3f}"
```
