# **1 | Understand**

<p><font color='black' size="5">
Anwendungsfall
</font></p>

In diesem Anwendungsfall wird der Apriori-Algorithmus eingesetzt, um hufig auftretende Kombinationen von Symptomen in Patientenbesuchen zu identifizieren. Der Datensatz besteht aus einer Liste von Patientenbesuchen, wobei jede innere Liste die whrend eines Besuchs gemeldeten Symptome enthlt.

Das Ziel ist es, Assoziationsregeln zu generieren, die zeigen, welche Symptomkombinationen hufig zusammen bei einem Patienten vorkommen. Beispielsweise knnte der Algorithmus feststellen, dass Patienten, die sowohl 'Fieber' als auch 'Husten' melden, mit hoher Wahrscheinlichkeit auch 'Mdigkeit' aufweisen.

Diese Analyse kann dazu beitragen, typische Symptommuster zu erkennen, potenzielle Zusammenhnge zwischen Symptomen aufzudecken und als Grundlage fr weitere Analysen oder die Entwicklung von Diagnosemodellen zu dienen. Es ist wichtig zu beachten, dass der Algorithmus lediglich Assoziationen findet und keine Kausalitt belegt.

```python
# Beispiel-Daten, die Patientenbesuche und gemeldete Symptome darstellt
# Jede innere Liste reprsentiert einen Patientenbesuch und die whrend dieses Besuchs gemeldeten Symptome
data = [
    ['Fieber', 'Husten', 'Halsschmerzen', 'Mdigkeit'],
    ['Fieber', 'Husten', 'Schnupfen', 'Mdigkeit'],
    ['Halsschmerzen', 'Kopfschmerzen', 'Mdigkeit'],
    ['Fieber', 'Husten', 'Halsschmerzen', 'Mdigkeit', 'Schnupfen'],
    ['Fieber', 'Kopfschmerzen', 'Mdigkeit'],
    ['Husten', 'Halsschmerzen', 'Schnupfen'],
    ['Fieber', 'Husten', 'Mdigkeit'],
    ['Halsschmerzen', 'Kopfschmerzen', 'Schnupfen', 'Mdigkeit'],
    ['Fieber', 'Halsschmerzen', 'Mdigkeit'],
    ['Husten', 'Schnupfen', 'Mdigkeit'],
    ['Fieber', 'Husten', 'Halsschmerzen', 'Mdigkeit'],
    ['Kopfschmerzen', 'Schnupfen', 'Mdigkeit'],
    ['Fieber', 'Husten', 'Schnupfen'],
    ['Halsschmerzen', 'Kopfschmerzen', 'Mdigkeit'],
    ['Fieber', 'Husten', 'Halsschmerzen', 'Mdigkeit']
]
```

<p><font color='black' size="5">
Parmeter-Set
</font></p>

```python
# bergreifende Parameter
min_support = 0.01
min_threshold = 0.60

# Spaltenauswahl
auszug = [
    "antecedents",
    "consequents",
    "antecedent support",
    "consequent support",
    "support",
    "confidence",
    "lift",
]
auszug_num = [
    "antecedent support",
    "consequent support",
    "support",
    "confidence",
    "lift",
]
auszug_confidence = [
    "antecedents",
    "consequents",
    "antecedent support",
    "consequent support",
    "support",
    "confidence",
]
auszug_lift = ["antecedents", "consequents", "lift"]
```

<p><font color='black' size="5">
Berechnung
</font></p>

```python
# Berechnung Frequenzen
frequent_itemsets = apriori(df, min_support=min_support, use_colnames=True)
frequent_itemsets.support = frequent_itemsets.support.round(2) * 100  # Umrechnung in %
frequent_itemsets = frequent_itemsets[["itemsets", "support"]]  # ndern Reihenfolge
```

```python
# Berechnung Regeln
associationRulesSet = association_rules(
    frequent_itemsets, num_itemsets=1, metric="confidence", min_threshold=min_threshold
)
associationRulesSet = associationRulesSet[auszug]
associationRulesSet[["confidence", "lift"]] = (
    associationRulesSet[["confidence", "lift"]] * 100
)
associationRulesSet[auszug_num] = associationRulesSet[auszug_num].round(2)
```

**Support (Hufigkeit)**   

Der Support sagt uns, wie hufig eine bestimmte Kombination von Symptomen in Patientenbesuchen vorkommt. Er wird als Prozentsatz aller Patientenbesuche angegeben.

`WENN man alle Patientenbesuche betrachtet, DANN ist in X% der Besuche Symptom A enthalten`

Beispiel: Wenn in 15 Patientenbesuchen 7-mal sowohl 'Fieber' als auch 'Husten' gemeldet wurden, dann ist der Support fr diese Kombination (7 / 15) * 100%  46,7%.

```python
frequent_itemsets.head(5)
```

```python
frequent_itemsets.tail(5)
```

**Confidence (Verlsslichkeit)**

Die Confidence zeigt, wie verlsslich eine Regel ist. Sie gibt an: Wenn Symptom A (antecedents) in einem Patientenbesuch beobachtet wird, wie wahrscheinlich ist es dann, dass auch Symptom B (consequents) ebenfalls beobachtet wird?

`WENN Symptom A in einem Patientenbesuch enthalten ist, DANN ist mit X% Wahrscheinlichkeit auch Symptom B in diesem Besuch enthalten`

Beispiel: Wenn von 15 Patienten, die 'Fieber' melden, 12 auch 'Husten' melden, dann ist die Confidence (12 / 15) * 100% = 80%. Die Regel "Wenn 'Fieber', dann 'Husten'" hat also eine hohe Verlsslichkeit in diesem Datensatz.

```python
associationRulesSet[auszug_confidence].head(5)
```

```python
associationRulesSet[auszug_confidence].tail(5)
```

```python
# Beispiel: Regeln basierend auf dem Konsequens (rechte Seite) filtern - Symptome identifizieren, die oft mit anderen verbunden sind
rules_mit_mdigkeit = associationRulesSet[associationRulesSet['consequents'].apply(lambda x: 'Mdigkeit' in x)]

print("Assoziationsregeln mit 'Mdigkeit' in der Konsequenz:")
rules_mit_mdigkeit[auszug_confidence].head(5)
```

```python
# Beispiel: Regeln basierend auf dem Antezedens (linke Seite) filtern - Symptomombinationen identifizieren, die oft zusammen auftreten
rules_mit_fieber_husten = associationRulesSet[associationRulesSet['antecedents'].apply(lambda x: 'Fieber' in x and 'Husten' in x)]

print("Assoziationsregeln mit 'Fieber' und 'Husten' im Antezedens:")
rules_mit_fieber_husten[auszug_confidence].head(5)
```

**Lift (Zusammenhang)**

Der Lift zeigt, ob zwei Symptome wirklich zusammenhngen oder ob sie nur zufllig zusammen auftreten.

+ Lift = 1  keine Beziehung, Symptome treten unabhngig voneinander auf

+ Lift > 1  eine positive Beziehung/Regel, Symptome treten hufiger zusammen auf

+ Lift < 1  negative Beziehung zwischen den Symptomen, gegenseitig ausschlieende oder substituierende Symptome

`WENN der Wert grer als 1 ist, DANN treten die Symptome A und B hufiger zusammen auf`

Beispiel: Ein Lift von 2,5 zwischen 'Fieber' und 'Husten' bedeutet, dass diese Kombination 2,5-mal hufiger in Patientenbesuchen vorkommt als es bei zuflligem Auftreten zu erwarten wre. Die Symptome haben also eine echte Verbindung zueinander in diesem Datensatz.

```python
associationRulesSet[auszug_lift].head(5)
```

```python
associationRulesSet[auszug_lift].tail(5)
```
