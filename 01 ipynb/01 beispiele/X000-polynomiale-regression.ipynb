```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
np.random.seed(0)

n_samples = 30
degrees = [1, 4, 13]

X = np.sort(np.random.rand(n_samples))
y = np.cos(1.5 * np.pi * X) + np.random.randn(n_samples) * 0.1  # Direkte Erzeugung der y-Werte

fig = make_subplots(rows=1, cols=3, subplot_titles=[f"Degree {degree}" for degree in degrees])

for i, degree in enumerate(degrees, start=1):
    polynomial_features = PolynomialFeatures(degree=degree, include_bias=False)
    X_transformed = polynomial_features.fit_transform(X[:, np.newaxis])

    linear_regression = LinearRegression()
    linear_regression.fit(X_transformed, y)

    X_test = np.linspace(0, 1, 100)
    X_test_transformed = polynomial_features.transform(X_test[:, np.newaxis])
    y_test = linear_regression.predict(X_test_transformed)

    fig.add_trace(go.Scatter(x=X, y=y, mode='markers', name='Samples', legendgroup=f"group{i}"), row=1, col=i)
    fig.add_trace(go.Scatter(x=X_test, y=y_test, mode='lines', name='Model', legendgroup=f"group{i}"), row=1, col=i)

fig.update_layout(height=400, width=1200, title_text="Polynomial Regression with Different Degrees")
fig.show()
```
