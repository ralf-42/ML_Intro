# **3 <font color='orange'>|</font> Modeling**
---



```python
def chat_with_openai(user_input):
    global chat_history

    # Nutzereingabe zum Chatverlauf hinzufügen
    chat_history.append({"role": "user", "content": user_input})

    # Sende den Chatverlauf an die OpenAI API und erhalte die Antwort
    # Use openai.chat.completions.create for Chat Completions API
    response = openai.chat.completions.create(
        model="gpt-4",  # Oder "gpt-3.5-turbo" wenn GPT-4 nicht verfügbar ist
        messages=chat_history
    )

    # Extrahiere die Antwort von OpenAI
    # Access the content using response.choices[0].message.content
    assistant_reply = response.choices[0].message.content

    # Antwort dem Chatverlauf hinzufügen
    chat_history.append({"role": "assistant", "content": assistant_reply})

    # Erstelle den gesamten Verlauf als Text
    chat_text = ""
    for message in chat_history:
        role = "User" if message["role"] == "user" else "Assistant"
        chat_text += f"{role}: {message['content']}\n\n"

    return chat_text
```

# **5 <font color='orange'>|</font> Deploy (Apply)**
---

```python
# Gradio Interface erstellen
with gr.Blocks() as demo:
    gr.Markdown("# OpenAI Chat über Gradio")

    # Ausgabefeld für den Chat-Verlauf (oben)
    chat_output = gr.Textbox(label="Chat Verlauf", placeholder="Der Chat-Verlauf wird hier angezeigt...", lines=20)

    # Eingabefeld für die Nutzerfrage (unten)
    user_input = gr.Textbox(label="Ihr Prompt:", placeholder="Geben Sie Ihren Prompt ein ...")

    # Button zum Senden
    submit_button = gr.Button("Senden")

    # Button-Aktion: Nutze die Funktion und aktualisiere den Chat-Verlauf
    submit_button.click(chat_with_openai, inputs=user_input, outputs=chat_output)
```

```python
# Initialisiere den Chat-Verlauf
chat_history = []
# Starte die Gradio-App
demo.launch()
```
