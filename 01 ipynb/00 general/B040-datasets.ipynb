Diese Arbeit entstand aus dem Wunsch, Gewebeproben ausschlielich auf der Grundlage einer Feinnadelpunktion (FNA) genau zu diagnostizieren. Der resultierende Datensatz ist als Wisconsin Breast Cancer Data bekannt.


Die Arbeit an der Bildanalyse begann 1990 mit der Aufnahme von Nick Street in das Forschungsteam. Ziel war es, die Probe anhand eines digitalen Bildes eines kleinen Abschnitts des FNA-Objekttrgers zu diagnostizieren.

**Diagnoseablauf:**

Es wird aus dem Gewebe eine FNA entnommen. Dieses Material wird dann auf einen Objekttrger montiert und gefrbt, um die Zellkerne hervorzuheben. Ein Teil des Objekttrgers, in dem die Zellen gut differenziert sind, wird dann mit einer Digitalkamera und einem Framegrabber-Board gescannt.
Anschlieend isoliert der Anwender die einzelnen Zellkerne . Mit einem Mauszeiger zeichnet der Benutzer die ungefhre Grenze jedes Kerns. Mithilfe eines Computer-Vision-Ansatzes, konvergieren diese Annherungen dann an die genauen nuklearen Grenzen. Sobald alle (oder die meisten) Kerne auf diese Weise isoliert wurden, berechnet das Programm Werte fr jedes der zehn Merkmale jedes Kerns und misst Gre, Form und Textur. Der Mittelwert, der Standardfehler und die Extremwerte dieser Merkmale werden berechnet, was zu insgesamt 30 Kernmerkmalen fr jede Probe fhrt.

Der Datensatz enthlt 9568 Datenpunkte, die von einem Gas- und Dampfturbinenkraftwerk ber einen Zeitraum von 6 Jahren (2006-2011) gesammelt wurden, als das Kraftwerk unter Volllast in Betrieb genommen wurde. Die Funktionen bestehen aus den stndlichen durchschnittlichen Umgebungsvariablen
+ Temperatur (T),
+ Umgebungsdruck (AP),
+ Relative Luftfeuchtigkeit (RH)
+ und Abgasvakuum (V),

um die stndliche Netto-Stromerzeugung (EP) der Anlage vorherzusagen.


Ein GuD-Kraftwerk (GuD-Kraftwerk) setzt sich aus Gasturbinen (GT), Dampfturbinen (ST) und Abhitzedampferzeugern zusammen. Bei einem GuD-Kraftwerk wird der Strom durch Gas- und Dampfturbinen erzeugt, die in einem Kreislauf kombiniert werden, und von einer Turbine auf eine andere bertragen. Whrend das Vakuum von der Dampfturbine beeinflusst wird und sich auf sie auswirkt, beeinflussen die anderen drei Umgebungsvariablen die GT-Leistung.

Der Hochdruckdampf in einem GuD-Kraftwerk wird im sogenannten Abhitzedampferzeuger (AHE) erzeugt  einem zentralen Bauteil, das die Abwrme der Gasturbine nutzt, um Wasser in Dampf umzuwandeln.

Nach der Expansion des Dampfes in der Dampfturbine muss dieser wieder kondensieren, um erneut im Kreislauf genutzt werden zu knnen. Das passiert im sogenannten Kondensator, einem Wrmebertrager, in dem der Dampf durch Khlwasser (z.B. Fluss- oder Meerwasser) abgekhlt wird.

Beim Kondensieren entsteht ein Unterdruck (Vakuum), weil Wasserdampf beim bergang in den flssigen Zustand viel Volumen verliert  und das senkt den Druck im Kondensator erheblich, typischerweise auf etwa 0,05 bar oder sogar weniger (je nach Khlwassertemperatur).

Das Vakuum erhht die Druckdifferenz ber der Dampfturbine, also:

`DruckamEinlass(Hochdruckdampf)DruckamAuslass(Vakuum)`


Je grer dieser Unterschied ist, desto mehr Energie kann der Dampf beim Expandieren in der Turbine abgeben.

Dadurch:

+ steigt die mechanische Leistung der Dampfturbine,

+ wird mehr Strom im Generator erzeugt,

+ und der Wirkungsgrad des gesamten GuD-Kraftwerks verbessert sich.

---

<p><font color='black' size="5">

Daten, die von der National Sports Promotion Foundation zum Gedenken an die Olympischen Spiele in Seoul verwaltet werden. Die Daten beinhatlten den Leistungsgrad, Altersgruppe, Gre, Gewicht, Sit-up, BMI und Weitsprung an Ort und Stelle, etc.  


[DataSet](https://www.kaggle.com/datasets/kukuroo3/body-performance-data)   
[Info](https://www.kaggle.com/datasets/kukuroo3/body-performance-data)   
[Quelle](https://www.bigdata-culture.kr/bigdata/user/data_market/detail.do?id=ace0aea7-5eee-48b9-b616-637365d665c1)



1. **Grip Force**: Bei dieser bung wird die Kraft gemessen, die eine Person mit ihrer Hand ausben kann, in der Regel mit einem Gert namens Dynamometer. Die Teilnehmer drcken das Gert so fest wie mglich zusammen. Die gemessene Kraft wird in Kilogramm oder Newton angegeben. **Negative Werte** sind hier nicht mglich, da die Kraftmessung immer einen positiven Wert oder null ergibt.

2. **Sit and Bend Forward_cm (Sitzen und nach vorne beugen in cm)**: Diese bung misst die Flexibilitt. Die Teilnehmer sitzen auf dem Boden mit gestreckten Beinen und beugen sich vorwrts, um so weit wie mglich mit den Hnden voranzukommen. Die Distanz, die sie erreichen knnen, wird gemessen. **Negative Werte** knnten theoretisch auftreten, wenn die Startposition als Nullpunkt definiert ist und jemand nicht ber diesen Punkt hinaus kommt. Allerdings ist das ungewhnlich und hngt von der spezifischen Messmethode ab.

3. **Sit-ups Counts (Anzahl der Sit-ups)**: Diese bung misst die Bauchmuskelkraft und Ausdauer. Die Teilnehmer fhren so viele Sit-ups wie mglich in einer festgelegten Zeit durch. Die Ergebnisse werden als Anzahl der vollstndig ausgefhrten Sit-ups gezhlt. **Negative Werte** sind hier nicht mglich, da es sich um eine Zhlung handelt und diese bei null beginnt.

4. **Broad Jump_cm (Weitsprung in cm)**: Bei dieser bung wird die explosive Beinkraft gemessen. Die Teilnehmer springen aus dem Stand so weit wie mglich nach vorne.
