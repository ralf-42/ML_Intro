```python
import os
from google.colab import userdata

from langchain.schema import HumanMessage, AIMessage, SystemMessage
from langchain.chains import LLMChain
import langchain.schema as me
from langchain_core.prompts import ChatPromptTemplate
from langchain_groq import ChatGroq

import gradio as gr
```

# **2 <font color='orange'>|</font>  App-Integration ChatBot**

---

```python
# Definiere die Systemnachricht und legt damit den Verhaltensrahmen für den virtuellen Assistenten fest.
system = "You are a helpful assistant. Please, be brief and concise and to the point. "
```

```python
# Definiere die Eingabevorlage für den Benutzer:
# 1. Den Gesprächsverlauf aus dem "Memory" (Speicher des Chatverlaufs) als Kontext.
# 2. Die aktuelle Benutzerabfrage (USER_QUERY), die beantwortet werden soll.
human = "Use the conversation memory below to help in answer the most recent query from the user:\n{memory}\n\n \
    USER_QUERY:{text}\n\n ASSISTANT:"
```

```python
# Erstelle eine Eingabeaufforderung (Prompt) aus den System- und Menschennachrichten:
# `ChatPromptTemplate.from_messages` erstellt eine kombinierte Eingabeaufforderung, die das Modell steuert.
prompt = ChatPromptTemplate.from_messages([("system", system), ("human", human)])
```

```python
# Initialisiere das Modell
chat = ChatGroq(temperature=0, model="llama3-70b-8192")
```

```python
# Konversationskette (conversation_chain):
# Die Eingabeaufforderung (prompt) wird hier mit dem Chat-Modell (chat) kombiniert.
# Das Ergebnis ist eine Kette, die den Verlauf und den aktuellen Kontext nutzt, um Antworten zu generieren.
conversation_chain = prompt | chat
```

```python
def respond(message, history):
    history_langchain_format = []
    for human, ai in history:
    history_langchain_format.append(me.HumanMessage(content=human))
    history_langchain_format.append(me.AIMessage(content=ai))
    history_langchain_format.append(me.HumanMessage(content=message))
    result = conversation_chain.invoke({"memory": history, "text":message})
    return result.content
```

# **3 <font color='orange'>|</font>  App Design**
---

```python
iface = gr.ChatInterface(respond)
```
