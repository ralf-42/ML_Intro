```python
# Import
import os

from langchain_groq import ChatGroq
from langchain.prompts import ChatPromptTemplate

from google.colab import userdata

import gradio as gr
```

# **1 <font color='orange'>|</font> Übergreifende Parameter**
***


```python
# Groq API Schlüssel setzen
GROQ_API_KEY= userdata.get('GROQ_API_KEY')
os.environ["GROQ_API_KEY"] = GROQ_API_KEY
```

# **2 <font color='orange'>|</font>  App-Integration ChatBot**

---

```python
# Initialisiere das Groq-Modell
llm = ChatGroq(
    model="mixtral-8x7b-32768",
    temperature=0.5,
    max_tokens=150
)
```

```python
# Definiere die Vorlagen für die Eingaben
prompt_template = ChatPromptTemplate.from_messages([
    ("system", "Du bist ein hilfreicher Assistent."),
    ("human", "{message}")
])
```

```python
# Chat mit Groq
def chat_groq(message, history):
    # Konvertiere den Verlauf in das Format der LangChain-API
    formatted_messages = [{"role": msg['role'], "content": msg['content']} for msg in history]
    formatted_messages.append({"role": "user", "content": message})

    # Erstelle eine Chat-Vervollständigung mit dem Groq-Modell
    chain = prompt_template | llm
    response = chain.invoke({"message": message})

    return response.content
```

# **3 <font color='orange'>|</font>  App Design**
---

```python
# Initialisiere das Gradio-Interface
demo = gr.ChatInterface(
    chat_groq,
    type="messages",
    title="🤖 Einfacher ChatBot",
)
```

# **4 <font color='orange'>|</font> Launch**
---

```python
demo.launch()
```
